# Folium : Créer des cartes

### Pour simplement créer une carte, il suffit de 3 lignes

```python
import folium
carte = folium.Map()
carte # Pour afficher la carte
```

### Centrer la carte

```python
import folium
carte = folium.Map(location=(48.856614,2.3522219),zoom_start=6) # location=(latitude,longitude) du centre de votre carte

carte.save("ma_carte.html") # Enregistre le résultat dans un fichier ma_carte.html dans le répertoire de travail
carte # Pour afficher la carte
```

```python
### Créer des cercles sur la carte
```

```python
import folium
carte = folium.Map(location=(48.856614,2.3522219),zoom_start=6) # location=(latitude,longitude) du centre de votre carte


folium.Circle(
    radius=100000,
    location=[44.837789,-0.57918],
    popup="Bordeaux",
    color="crimson",
    fill=False,
).add_to(carte)
folium.Circle(
    radius=100000,
    location=[49.258329,4.031696],
    popup="Reims",
    color="blue",
    fill=True, # Remplissage
).add_to(carte)
carte # Pour afficher la carte
```

Plutôt que d'écrire deux fois folium.Circle utilisez une boucle qui recupère les informations dans un fichier CSV

### Créer des marques

```python
import folium
carte = folium.Map(location=(48.856614,2.3522219),zoom_start=6) # location=(latitude,longitude) du centre de votre carte

folium.Marker(
    location=[44.837789,-0.57918],
    popup="Bordeaux",
    icon=folium.Icon(icon="cloud"),
).add_to(carte)
folium.Marker(
    location=[49.258329,4.031696],
    popup="Reims",
    icon=folium.Icon(icon="cloud",color="red"),
    color="blue"
).add_to(carte)
carte
```

Il existe plusieurs types d'icones comme icon="cloud" ou icon="info-sign"
