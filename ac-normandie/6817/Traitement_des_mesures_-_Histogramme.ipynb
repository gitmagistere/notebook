```python
import matplotlib.pyplot as plt
from numpy import loadtxt
from statistics import mean, stdev  # Import des fonctions moyenne (mean) et écart-type (stdev)

"""Importation des données depuis le fichier mesures.csv
et stockage dans une liste appelée D
NE PAS MODIFIER"""

D = loadtxt("mesures.csv", skiprows=0,  dtype="str", delimiter=";", unpack=True, usecols=(0))
D = [float(val.replace(",", ".")) for val in D]

"""Création du graphique
-----------------------------------------------------------------------"""
plt.clf()  # Réinitialisation de la fenêtre graphique
plt.figure(1)
plt.title("ENTRER LE TITRE DE VOTRE GRAPHIQUE")  # Titre du graphique
plt.hist(D, bins=25, color="blue")  # Ajuster le paramètre 'bins' pour avoir plus ou moins de classes
plt.grid()  # Affichage de la grille
plt.xlabel("GRANDEUR MESURÉE (UNITÉ)")  # Entrez la légende de l'axe des abscisses avec la grandeur mesurée et son unité.
plt.show()  # Affichage du graphique

"""Calculs statistiques"""
"""-----------------------"""
print("Moyenne : ", mean(D))
print("Écart-type de la série : ", stdev(D))
print("Nombre de valeurs : ", len(D))
print("Incertitude type sur la moyenne : ", stdev(D)/(len(D))**0.5)
print("Valeur minimale : ", min(D))
print("Valeur maximale : ", max(D))
print("Étendue : ", max(D) - min(D))
```
