** de la mesure d'un volume**

```python
from matplotlib import pyplot as plt
import statistics as stat

# volume de 100 mL mesurés avec un bécher
V=[95.00, 95.32, 95.63, 96.42, 96.88, 97.22, 97.47, 97.68, 97.71, 97.85, 97.88, 98.11, 98.38, 98.63, 99.09, 99.45, 99.65, 99.87, 99.96, 100.29, 100.32, 101.83]

#calculs des caractéristiques statistiques et affichage dans la console
moyenne = stat.mean(V)
ecartype = stat.stdev(V)
print("les mesures ont pour moyenne V =",moyenne,'mL et pour écart-type expériemntal', ecartype, 'mL')

# Tracé de l'histogramme
plt.hist(V, range = (95,102), bins= 10, color = 'blue', edgecolor = 'black')
plt.xlabel("volumes en mL")
plt.ylabel("fréquences")
plt.title("Histogramme des notes")
plt.show()
```
