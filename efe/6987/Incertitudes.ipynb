#Simulation de la mesure d'un volume de 100 mL avec un bécher

```python
import numpy as np
import matplotlib.pyplot as plt
```

#création d'un tableau de 10000 valeurs issues d'une loi normale de moyenne 98.21 et d'écart type 1.76

```python
data = np.random.normal(98.21,1.76,10000)
```

#Tracé de l'histogramme

```python
plt.hist(data, bins=100, normed=1,color="lightblue",edgecolor = 'black')
```

Tracé de l'histogramme

```python
plt.title('Simulation de mesures aléatoire de volume \n en utilisant une loi normale de distribution')
plt.grid()
plt.savefig("loi de distribution de mesure de volume.png", bbox_inches='tight')
plt.show()
```
