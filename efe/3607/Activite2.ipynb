## Modélisation de la trajectoire d'un ballon
La trajectoire d'un ballon de basket....

```
import csv
import requests
import numpy as np
import matplotlib.pyplot as plt

source = requests.get('http://www.acamus.net/basket2.csv') 
source=source.content.decode('utf-8') 
source=source.splitlines()
t,x, y = [], [], []
for row in csv.reader(source,delimiter=';'):
  t1,x1,y1 = map(float,row) 
  t.append(t1)
  x.append(x1)
  y.append(y1)


n=np.arange(len(t)-1)
vyl=[]
vxl=[]
for i in n :
  vx=(x[i+1]-x[i])/(t[i+1]-t[i])
  vy=(y[i+1]-y[i])/(t[i+1]-t[i])
  vxl.append(vx)
  vyl.append(vy)
 

mod=np.polyfit(x,y,2)
print('mod :',mod)

m=np.arange(len(t))
ymod=[]
for i in m :  
  y2=mod[0]*x[i]**2+mod[1]*x[i]+mod[2]
  ymod.append(y2)

                   
plt.xlabel('abscisse (en m)')   
plt.ylabel('ordonnée (en m)')  
plt.grid()          
plt.title('Trajectoire du ballon') 
for i in n :
  plt.quiver(x[i],y[i],vxl[i], vyl[i],angles='xy', scale=25, scale_units='xy')
plt.show() 

plt.plot(x,y,'r+') 
plt.plot(x,ymod,'g-',label='modèle') 
plt.xlabel('abscisse (en m)')   
plt.ylabel('ordonnée (en m)')  
plt.grid()          
plt.title('Trajectoire du ballon') 
plt.legend() 
plt.show() 
```
