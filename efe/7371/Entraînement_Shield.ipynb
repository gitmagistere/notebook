## **Epreuve Shield 3 : Afficher la valeur de tous les pixels d’une image.**

1- Exécuter le programme ci-dessous en cliquant sur le petit triangle "Play" :

```
from PIL import Image
img = Image.open("pomme.jpg")
largeur_image=500
hauteur_image=500
for y in range(hauteur_image):
	for x in range(largeur_image):
    	      r,v,b=img.getpixel((x,y))
    	      print("pixel(",x,",",y,") --> rouge : ",r,", vert : ",v,", bleu : ",b)
print("fin")
```

On voit que ce programme affiche successivement les valeurs R,V et B de chaque pixel de l'image ligne par ligne.

## **Epreuve Shield 4 : Modifier la valeur de tous les pixels d'une image dans une dominante de couleur !**

### A] En programmant ...

1- Exécuter le programme ci-dessous en cliquant sur le petit triangle "Play" :

```
from PIL import Image
img = Image.open("pomme.jpg")
largeur_image=500
hauteur_image=500
for y in range(hauteur_image):
    for x in range(largeur_image):
        r,v,b=img.getpixel((x,y))
        n_r=r
        n_v=v
        n_b=255
        img.putpixel((x,y),(n_r,n_v,n_b))
img.save("pommebleue.bmp")
```

2- Cliquez sur le bouton "Actualiser" dans le menu de gauche et ouvrez le fichier "pommebleue.jpg".

On constate que ce programme change la couleur de la pomme en bleue. Le principe est simple, il suffit d'attribuer la valeur maximale au canal bleu du pixel, soit 255.

## **Epreuve Shield 5 : Convertir une image en niveau de gris !**

```
from PIL import Image
img = Image.open("pomme.jpg")
largeur_image=500
hauteur_image=500
for y in range(hauteur_image):
    for x in range(largeur_image):
        r,v,b=img.getpixel((x,y))
        n_r=v
        n_v=v
        n_b=v
        img.putpixel((x,y),(n_r,n_v,n_b))
img.save("pommeniveaugris.bmp")
```

2- Cliquez sur le bouton "Actualiser" dans le menu de gauche et ouvrez le fichier "pommeniveaugris.jpg".

On constate que ce programme change la couleur de la pomme en niveaux de gris. Le principe est simple, il suffit de d'attribuer la même valeur à chaque canal. Communément on utilise le vert qui est la couleur que nous perçevons le plus à l'oeil nu !!!
