```python
# On considère la réaction a A + b B -> c C + d D support d'un titrage

# A est le réactif titrant et B est le réactif titré

# CA et CB ont les concentrations respectives des réactifs A et B, VB est le volume de réactif titré, VA correspond au volume de solution de solution titrante versé, variant de 0 à 25,0 mL

# Ce programme vise à tracer les courbes donnant l'évolution des quantités de matière des réactifs et des produits à l'état final en fonction du volume VA de solution titrante versé

# Importation de la bibliothèque pour tracer des courbes
from matplotlib import pyplot as plt 

# Affectation des coefficients stoechiométriques
a = 1
b = 2
c = 2
d = 1


# Affectation des valeurs aux grandeurs constantes
CA = 0.10
CB = 0.10
VB = 20.0

# Création de listes pour les grandeurs variables avec  1 élément dans la liste correspondant à la condition initiale
VA = [0]
nA_f = [0]
nB_f = [CB*VB]
nC_f = [0]
nD_f = [0]

#Définition du pas pour le volume de solution titrante
pasVA = 0.10

#Initialisation du compteur
i = 0

for i in range(250) :
  i=i+1
  VA.append (VA[-1]+pasVA)
  if CA*VA[i]/a < CB*VB/b :
    nA_f.append (0)
    nB_f.append (CB*VB - b*CA*VA[i]/a)
    nC_f.append (c*CA*VA[i]/a)
    nD_f.append (d*CA*VA[i]/a)
  else :
    nA_f.append (CA*VA[i] - a*CB*VB/b)
    nB_f.append (0)
    nC_f.append (c*CB*VB/b)
    nD_f.append (d*CB*VB/b)

plt.xlabel('Volume VA de solution titrante versé (mL)')
plt.ylabel('Quantites de matiere (mmol)')
plt.title('Evolution des quantites de matiere des réactifs et des produits au cours d un titrage')
plt.grid()
plt.plot(VA, nA_f,color = 'red', label='Quantite nA_f')
plt.plot(VA, nB_f,color = 'blue', label='Quantite nB_f')
plt.plot(VA, nC_f,color = 'yellow', label='Quantite nC_f')
plt.plot(VA, nD_f,"--", color = 'orange', label='Quantite nD_f')
plt.legend()
plt.show()


```
