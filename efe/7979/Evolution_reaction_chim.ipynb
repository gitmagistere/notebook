##Evolution des quantités de matières
###Equation:  I2(aq) + 2S2O32-(aq) --> 2I-(aq)+ S4O62-(aq)

```python
import matplotlib.pyplot as plt
```

Valeurs initiales à définir

```python
a, b, c, d = 1, 2, 2, 1 # coefficients stoechiométriques
n_diiode, n_thiosulfate, n_iodure, n_tetrathionate = 1e-4,4e-4, 0, 0 # Quantités de matière initiales
n_diiode, n_thiosulfate, n_iodure, n_tetrathionate = [n_diiode], [n_thiosulfate], [n_iodure], [n_tetrathionate] # Initialisation des listes de valeurs
x = [0] #valeur initiale de l'avancement
dx = 1e-6   #valeur du pas
```

Boucle while pour calculer les contenus des listes

```python
while (n_diiode[-1]>0) and (n_thiosulfate[-1]>0) :
    x.append(x[-1] + dx)            #la dernière valeur de x est augmentée de dx
    n_diiode.append(n_diiode[-1] - a*dx)  #la dernière valeur de n_diiode est diminuée de a*dx
    n_thiosulfate.append(n_thiosulfate[-1] - b*dx)
    n_iodure.append(n_iodure[-1] + c*dx)
    n_tetrathionate.append(n_tetrathionate[-1] + d*dx)
    xMax = x[-1]
```

Représentation graphique

```python
plt.plot(x,n_diiode,label = "Réactif diiode")
plt.plot(x,n_thiosulfate,label = "Réactif thiosulfate")
plt.plot(x,n_iodure,label = "Produit iodure")
plt.plot(x,n_tetrathionate,label = "Produit tetrathionate")
plt.title("Evolution d'une réaction chimique")
plt.xlabel("Avancement x (mol)")
plt.xlim(0,xMax)
plt.ylabel("Quantité de matière (mol)")
plt.ylim(0,4e-4)
plt.legend()
plt.grid()
plt.show()
```
