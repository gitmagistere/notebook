 # Boucle for

Une **boucle** permet de répéter des instructions. Lorsqu'on connaît à l'avance le nombre de répétitions, on parle de **boucle bornée**. L'instruction <code>for</code> (pour) permet  de créer une boucle bornée.


En langage Python, la syntaxe d'une boucle for est la suivante :



```python
for i in range(...):
    instruction(s)
```

**Remarques**

* La fonction <code>range()</code> permet d'énumérer le nombre de passages dans la boucle bornée. Elle peut être appelée de différentes façons :

1) <code>range(n)</code>, où <code>n</code> est un entier, fait prendre à la variable <code>i</code> les valeurs entières de 0 à n-1, donc n valeurs ;

2) <code>range(n,m)</code>, où <code>n</code> et <code>m</code> sont des entiers, fait prendre à la variable <code>i</code> les valeurs entières de n à m-1 ;

3) <code>range(n,m,k)</code>, où <code>n</code>, <code>m</code> et <code>k</code> sont des entiers, fait prendre à la variable <code>i</code> les valeurs entières de n à m-1 avec un pas de k.

* La variable <code>i</code> n'est pas forcément utilisée dans les instructions à répéter, on l'appelle **compteur**.

* Les instructions à répéter sont indentées : il n'y a pas d'instructions pour signaler la fin de la boucle, c'est l'indentation qui indique les instructions faisant partie de la boucle.

<div class="alert alert-info"><strong>Exercice 4 : </strong> On considère la fonction ci-dessous dont l'argument <code>a</code> est un nombre entier naturel non nul :
</div>

```python
def nbdiv(a):
    n=0
    for i in range(1,a+1):
        if a%i==0:
            n=n+1
    return n
```
