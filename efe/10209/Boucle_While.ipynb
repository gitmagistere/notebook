# Boucle while

<p style="text-align:justify;">La boucle <code>for</code> a l'avantage d'être simple à utiliser, mais elle nécessite de connaître à l'avance (avant d'entrer dans la boucle) le nombre de tours de boucles qui vont être effectués. Dans certaines situations, on souhaiterait demander au programme de sortir de la boucle lorsqu'une certaine condition est réalisée, par exemple lorsqu'une valeur dépasse un certain seuil, ou lorsque l'utilisateur a saisi le bon mot de passe ou cliqué sur un bouton. La boucle <code>while</code> permet de répondre à ce besoin.</p>

**Syntaxe de la boucle while** <br><br>
<p style="border:solid;padding:5px">
   <code>while condition:
    bloc d'instructions A       
bloc d'instructions B
</code></p>

<p style="text-align:justify;">Tant que la condition est vraie, le bloc d'instructions A est exécuté, généralement plusieurs fois. Les instructions du bloc A agissent sur la condition pour la rendre fausse au bout d'un certain nombre de tours. Alors le programme sort de la boucle et se poursuit  avec l'exécution du bloc B.</p>

<strong>Remarques : </strong>
<ul style="text-align:justify;">
    <li>
La boucle <code>while</code> est parfois appelée "boucle non bornée" (nombre de tours non connu à l'avance, pouvant éventuellement dépasser n'importe quelle valeur) par opposition à la boucle <code>for</code> qui est appelée "boucle bornée" (le nombre de tours de boucle ne dépassera pas une valeur fixée à l'avance).
    </li>
    <li>
Il est possible que la boucle ne soit jamais exécutée, si la condition est initialement fausse.
    </li>
    <li>
       Il est possible que le programme boucle indéfiniment et ne s'arrête jamais, lorsque la condition est toujours vraie. On parle alors de <strong>boucle infinie</strong>. <br>
