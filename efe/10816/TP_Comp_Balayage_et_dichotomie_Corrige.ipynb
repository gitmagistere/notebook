## 1. Algorithme de balayage   <a name="1"></a> 

Compléter la fonction Python `balayage()` de paramètres `debut`, `fin` (qui représentent le début et la fin de l’intervalle où l’équation $f(x) = 0$ admet une solution)  et `précision ` (l’amplitude de l’encadrement que l’on souhaite pour la solution).

Besoin d'aide? Cliquez ici.

<div style="border-left:3px solid #3c763d; border-radius:2pt; color:#3c763d; padding:6px; background-color: #eff0e8">

On doit à chaque fois tester si <code> f(debut) </code> et <code> f(debut+precision) </code> ont le même signe. 
    <br>S'ils ont le même signe, la solution ne se trouve pas entre <code> debut </code> et <code> debut+precision </code>, il faut donc changer de <code> debut </code> qui prend pour valeur <code> debut+precision </code>.
    <br>Par contre dès que <code> f(debut) </code> et <code> f(debut+precision) </code> sont de signes contraires, on a gagné, la solution se trouve entre <code> debut </code> et <code> debut+precision </code>.
    <br>Il reste le plus important : <b> Deux nombres ont le même signe lorsque leur produit est positif  </b>.
</div>
</details>

```python
def f(x):
    return x**3 - 10

def balayage(debut, fin, precision):
    while f(debut) * f(debut + precision) > 0:   
        debut = debut + precision
    return 'solution entre', debut,'et', debut + precision

       
```

```python
balayage(0, 3, 0.1) #tester votre fonction une fois complétée
```

## 2. Algorithme de dichotomie <a name="2"></a> 

Compléter la fonction Python `dichotomie()` de paramètres `debut`, `fin` (qui représentent le début et la fin de l’intervalle où l’équation $f(x) = 0$ admet une solution)  et `précision ` (l’amplitude de l’encadrement que l’on souhaite pour la solution).

```python
def f(x):
    return x**3 - 10

def dichotomie(debut, fin, precision):
    while fin - debut > precision:  # On teste si l ecart entre debut et fin est superieur a precision 
        x = (debut+fin)/2  # x doit prendre pour valeur le milieu de [debut, fin]
        if f(debut) * f(x) > 0 :  # On teste si f(debut) et f(x) ont meme signe
            debut = x
        else:
            fin = x
    return 'solution entre', debut,'et', fin
```

```python
dichotomie (0, 3, 0.1)  # Tester votre fonction une fois complétée
```

```python
dichotomie (0, 3, 0.01)
```

## 3. Comparatif balayage et dichotomie  <a name="3"></a> 
Compléter la fonction Python `nombre_calculs_balayage()` pour qu'elle renvoie le nombre de calculs nécessaires (nombre de tests) pour déterminer un encadrement d'amplitude donné (`précision`) de la solution de l'équation $f(x) = 0$ sur l'intervalle [`debut`, `fin`] par la méthode de balayage.  
Compléter la fonction Python `nombre_calculs_dichotomie()` pour qu'elle renvoie le nombre de calculs nécessaires (nombre de tests) pour déterminer un encadrement d'amplitude donné (`précision`) de la solution de l'équation $f(x) = 0$ sur l'intervalle [`debut`, `fin`] par la méthode de dichotomie.

```python
def f(x):
    return x**3 - 10

def nombre_calculs_balayage(debut, fin, precision):
    compteur = 1  # On initialise le compteur par 1 car on va faire un test
    while f(debut) * f(debut + precision) > 0:
        debut = debut + precision
        compteur += 1  # On augmente le compteur de 1 car on va faire un test
    return compteur

def nombre_calculs_dichotomie(debut, fin, precision):
    compteur = 0
    while fin - debut > precision :  # On teste si l ecart entre debut et fin est superieur a precision 
        x = (debut+fin)/2  # x doit prendre pour valeur le milieu de [debut, fin]
        compteur += 1 # On augmente le compteur de 1 car on va faire un test
        if f(debut) * f(x) > 0 :  # On teste si f(debut) et f(x) ont meme signe
            debut = x
        else:
            fin = x
    return compteur


def comparatif_balayage_dichotomie(debut, fin, precision):
    return f"Il faut {nombre_calculs_balayage(debut, fin, precision)} calculs avec la méthode balayage tandis qu'avec la méthode dichotomie il ne faut que {nombre_calculs_dichotomie(debut, fin, precision)} calculs pour déterminer l'encadrement avec une précision de {precision}"

comparatif_balayage_dichotomie(0, 3, 0.1)
```
