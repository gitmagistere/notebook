# 1- Définition de la boucle non bornée "while"

Dans un certain nombre de cas, il est indispensable de répéter des instructions sans savoir à l'avance combien de fois on les répète. La boucle bornée "for" est alors inadaptée et on utilise un autre type de boucle, une **boucle non bornée**. Les instructions sont alors répétées tant qu'une condition est vérifiée. L'instruction <code>while</code> (tant que) permet  de créer une boucle non bornée.


En langage Python, la syntaxe d'une boucle while est la suivante :

```python
while condition:
    instruction(s)
```

**Remarque**

Les instructions à répéter sont indentées : il n'y a pas d'instructions pour signaler la fin de la boucle, c'est l'indentation qui indique les instructions faisant partie de la boucle.

**Exemple**

Dans le programme ci-dessous, la valeur de la variable <code>a</code> est multipliée par 3 tant qu'elle est strictement inférieure à 8. 

La variable <code>a</code> vaut donc successivement 1 , 3 et 9 avant l'arrêt de la boucle.

```python
a=1
while a<8:
    a=a*3
```

<div class="alert alert-info"><strong>Exercice 1 : </strong> Un client place le 1er janvier 2022 une somme de 2 000 € sur un compte bancaire. Le compte rapporte de l'argent chaque année en appliquant un taux de 2 %. Le client attend que la somme d'argent sur le compte dépasse 2 700 € pour retirer la totalité de l'argent. On considère le script suivant :
</div>

```python
def compte():
    somme=2000
    n=0
    while somme<=2700:
        somme=1.02*somme
        n=n+1
    return 2022+n
```

**Remarques**

* La variable <code>n</code> est un compteur : elle est initialisée à 0 et permet de compter le nombre de fois où l'on va effectuer l'augmentation.

* La fonction <code>compte</code> n'a pas d'argument.

<div class="alert alert-info">Exécuter l'instruction ci-dessous et interpréter le résultat renvoyé par la fonction <code>compte</code>.
</div>

```python
compte()
```

<div class="alert alert-info"><strong>Exercice 2 : </strong> Camille a écrit le script d'une fonction <code>pgm</code> dont les arguments sont deux entiers naturels <code>a</code> et <code>b</code> pour déterminer le plus grand multiple de <code>a</code> inférieur ou égal à l'entier <code>b</code>.

```python
def pgm(a,b):
    x=0
    while x<=b:
        x=x+a
    return x
```

<div class="alert alert-info">La fonction <code>pgm</code> permet-elle d'obtenir le résultat attendu ? Pourquoi ?

**Réponse** :<b> ce programme ne permet pas de répondre à la question, il affiche un multiple de a qui est plus grand que b !</b>

<div class="alert alert-info">Réécrire ci-dessous le script de la fonction <code>pgm</code> en le corrigeant.

```python
def pgm(a,b):
    x=0
    while x<=b:
        x=x+a
    return x-a
```

<div class="alert alert-info"><strong>Exercice 3 : </strong> La grand-mère de Solal lui a donné 50 € pour son anniversaire qu'il met dans sa tirelire. Ses parents lui donnent 15 € d'argent  de poche chaque mois que Solal décide déconomiser dans sa tirelire jusqu'à ce qu'il puisse s'acheter la guitare dont il rève et qui coûte 490 €. 
</div>

<div class="alert alert-info">Compléter le script de la fonction <code>guitare</code> ci-dessous afin qu'elle renvoie le nombre de mois que devra attendre Solal avant de s'acheter sa guitare.
</div>

```python
def guitare():
    n=0
    tirelire=50
    while tirelire < 490 :
        tirelire = tirelire + 15
        n = n+1
    return n
```
