## une naissance
La bibliothèque `random` de Python permet d'importer la fonction `random()` qui renvoie un nombre aléatoire de l'intervalle [0 ; 1].
<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

-  **Exécutez** la cellule ci-dessous :
</div>


```python
from random import random
for i in range (10):
    print(random())
```

On souhaite créer une fonction aléatoire qui renvoie *vrai* (`True` en python) si la naissance est une fille et *faux* `False` sinon.

<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Complétez** les pointillés de cette première fonction pour qu'elle simule une naissance, puis exécutez la cellule :</div>

```python
def naissance():
    return (random() < 0.5)  # le résultat d'un test est vrai ou faux
```

On ne voit rien car python prépare la fonction en mémoire.


<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Appelons** 10 fois cette fonction pour vérifier.</div>

```python
for i in range(10):
    print(naissance())
```

Comme notre fonction naissance semble satisfaisante, fabriquons une fonction famille qui renvoie le nombre de filles après 2 naissances.
## une famille
<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Complétez** les pointillés de cette seconde fonction pour qu'elle renvoie le **nombre de filles** après 2 naissances au sein d'une famille, puis exécutez la cellule :</div>

```python
def famille():
    nombre = 0 # il n'y a pas encore eu de naissance
    for n in range(2) : # on répète 2 simulations de naissance
        if naissance() : # naissance est vrai lorsqu'on a une fille
            nombre = nombre + 1
    return nombre
```

On ne voit rien car python prépare la fonction en mémoire.


<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Appelons** 10 fois cette fonction pour commencer à nous faire une idée.</div>

```python
for i in range (10):
    print (famille())
```

Comme notre fonction `famille()` semble satisfaisante, simulons beaucoup de familles ayant 2 naissances et calculons la proportion de 2 filles.
## un échantillon de familles
<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Complétez** les pointillés de cette troisième fonction pour qu'elle renvoie la proportion de familles simulées ayant 2 filles parmi n famille ayant 2 enfants, puis exécutez la cellule :</div>

```python
def echantillon(n):
    compte = 0  # on n'a pas encore regardé de famille
    for taille in range(n):
        if famille() == 2 :
            compte = compte + 1
    return compte/n        
```

On ne voit rien car python prépare la fonction en mémoire.


<div class="alert alert-info" style="border-left:5px solid #31708f;border-radius: 5px;">

- **Appelons** 5 fois cette fonction pour voir.</div>

```python
for i in range(5):
    print (echantillon(1000)) # On simule un échantillon de 1000 familles
```
