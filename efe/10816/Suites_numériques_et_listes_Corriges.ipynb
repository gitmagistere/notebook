<h3> PARTIE 1 : Calculer les termes d'une suite numérique avec une formule explicite </h3>
</div> 

#### Exemple : 
Soit $(u_n)$ la suite numérique définie pour tout $n\in \mathbb{N}$ par : $u_n=\frac {n-4}{2n+1}$. <br>
La fonction <code>suite_u(n)</code> ci-dessous envoie en sortie le n-ième terme pour un entier $n\in \mathbb{N}$ passé en argument. <br>
On peut alors calculer le terme $u_{20}$. <br>
N'oubliez pas d'exécuter le code Python de la cellule contenant la fonction <code>suite_u(n)</code>.

```python
def suite_u(n):
    return (n-4)/(2*n+1)
```

```python
suite_u(20)
```

<h3> PARTIE 2 : Calculer les termes d'une suite numérique avec une formule récurrente </h3>
</div> 

#### Exemple : 
Soit $(w_n)$ la suite numérique définie par $w_0=5$ et pour tout $n\in \mathbb{N}$ par : $w_{n+1}=3w_n+6$.
La fonction suivante renvoie le n-ième terme pour un entier $n\in \mathbb{N}$ passé en argument.

```python
def suite_w(n):
    w=5
    for i in range(n):
        w=3*w+6
    return w
```

### Appel de la fonction
Pour calculer $w_{10}$, on appelle la fonction `suite_w` avec l'argument 10.

```python
suite_w(10)
```

            et pour tout $n\in \mathbb{N}$ par : $t_{n+1}=0,5t_n+4$.</li>

**Qu'est ce qu'une liste en Python ?**

Une liste est une séquence ordonnée de variables. Chaque variable de la 
liste est identifiée par sa position et peut stocker une valeur
spécifique. Les éléments d'une liste s'écrivent entre crochets séparés 
par une virgule.

**La compréhension de liste**

La compréhension de liste est une technique propre à Python permettant de créer des listes de manière simple et efficace. Par exemple, pour créer la liste des 20 premiers entiers pairs, on peut écrire :

```python
pairs = [2*i for i in range(20)] 
print(pairs)
```
