### Dans ce TP, on cherche à calculer une approximation de l'aire sous une courbe.

La courbe est la représentation graphique d'une *fonction positive $f$ et croissante sur un intervalle*.

1. Définition de la fonction f :

```python
def f(x):
    return 1/x
```

2. `rectG()`doit calculer une approximation de $\int_0^1 f\,\text{d}x$ à l'aide de 5 rectangles à gauche.


```python
def rectG():
    S=0
    a=0
    b=1
    n=10
    pas = (b-a)/n
    for i in range(n):
        g=a+i*pas
        S = S + f(g)*pas        
    return S
rectG()
```

3. `rectD()`doit calculer une approximation de $\int_0^1 f\,\text{d}x$ à l'aide de 5 rectangles à droite.


```python
def rectD():
    S=0
    a=0
    b=1
    n=10
    pas = (b-a)/n
    for i in range(1,n+1):
        d=a+i*pas
        S = S + f(d)*pas        
    return S

rectD()
```

```python
def rectG(n):
    S=0
    a=0
    b=1
    pas = (b-a)/n
    for i in range(n):
        g=a+i*pas
        S = S + f(g)*pas        
    return S
rectG(100000)
```

5. `rectD(n)`doit calculer une approximation de $\int_0^1 f\,\text{d}x$ à l'aide de $n$ rectangles à droite.

```python
def rectD(n):
    S=0
    a=0
    b=1
    pas = (b-a)/n
    for i in range(1,n+1):
        d=a+i*pas
        S = S + f(d)*pas        
    return S

rectD(100)
```
