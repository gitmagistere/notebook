### 3. Comptage du nombre de piles au cours de n lancers de dé<a name="pile"></a>
On considère la fonction `nombre_de_piles(n)` définie dans le script ci-dessous.  
3.a. Compléter la fonction `nombre_de_piles(n)` pour qu'elle renvoie le nombre de 'Pile' obtenus lorsque l'on lance n pièces.  
3.b. Tester votre fonction avec 10 pièces puis 500 pièces.

```python
from random  import randint
def nombre_de_piles(n):
    piece = ['pile', 'face']
    compteur = 0  # On initialise compteur qui va compter le nombre de piles obtenus, 
    for i in range(n):  # On repete n fois 
        lancer = piece[randint(0, 1)]  # On simule le lancer d une pièce
        if lancer == piece[0] :  # Ici on doit vérifier si le lancer donne 'pile'
            compteur = compteur + 1  # Si on a obtenu 'pile', le nombre de piles augmente de 1
    return compteur
```

```python
nombre_de_piles(10)
```

### 1. Simulation <a name="sim"></a>  
Une urne contient 2 boules rouges et 3 boules noires.  
On tire successivement 2 boules de l'urne (sans remettre la première boule dans l'urne) et chaque boule rouge tirée rapporte 5 euros.  
On doit payer 2 euros pour participer à ce jeu.  
Ecrire ci-dessous une fonction `jeu()` qui simule ce jeu et renvoie le gain algébrique du joueur.

```python
from random  import randint
def jeu():
    gain = -2
    urne = ['rouge', 'rouge', 'noir', 'noir', 'noir']
    liste  = []
    for i in range(2):
        tirage = urne[randint(0, 3-i)]  # On simule ici le tirage d une boule dans l urne en tenant compte qu'à chaque fois il y a une boule en moins
        liste.append(tirage)  # Le tirage est ajoute dans liste
        urne.remove(tirage)  # On retire a l urne le resultat du tirage
    for i in range(2):
        if liste[i]=='rouge':
            gain+=5
            
    return gain
```

### 2. Estimation du gain moyen <a name="moy"></a>  
Ecrire ci-dessous une fonction `gain_moyen(n)` qui simule n parties du jeu précédent et renvoie le gain moyen.  
Tester votre fonction pour les valeurs $n=10$, $n=1000$ et $n=100000$.

```python
def gain_moyen(n):
    gain=0
    for i in range(n):
        gain+=jeu()
    return gain
gain_moyen(10)
```

### 3. Calcul de l'espérance mathématique <a name="esp"></a>  
3.a. Ecrire ci-dessous une fonction `esperance(liste_valeurs, liste_proba)` qui renvoie l'espérance mathématique d'une variable aléatoire X lorsque l'on entre la liste des valeurs prises par X et les probabilités correspondantes.  
3.b. Tester votre fonction avec la variable aléatoire ayant la loi de probabilité suivante :  

|  X=xi       | 0|50  |200| 400  |
| ------------- |:-------------:| :-----:|:-------------:| :-----:|
| P(X=xi)          | 993/1000       |  5/1000     |1/1000       |    1/1000   |

3.c. Tester votre fonction avec la variable aléatoire de l'exemple du 1



```python
def esperance(liste_valeurs, liste_proba):
    E=0
    for i in range(len(liste_valeurs)):
        E+=liste_valeurs[i]*liste_proba[i]
    return E
esperance([1,2], [0.5,0.5])
```

### 5. Qui veut gagner des millions? <a name="qui"></a>  

Vous devez participer à un jeu d'argent et vous avez le choix entre les deux jeux suivants.

#### Le jeu des dix dés
Ce jeu consiste à lancer dix dés et on gagne six euros pour chaque nombre 6 obtenu.
On peut donc gagner 0 euro (si on n'a pas de chance), 6 euros (si seulement un des dix dés est un 6), 12 euros (si on obtient deux 6) ... et jusqu'à 60 euros (si on est très chanceux).

#### Le jeu "jusqu'à l'infini et au-delà"
Ce jeu consiste à lancer un dé jusqu'à l'obtention d'un 6. On gagne un euro par lancer.
On peut donc gagner 1 euro (si on 6 dès le premier lancer), 2 euros (si on a au premier lancer 1-2-3-4 ou 5 et au deuxième lancer 6), ....,  125 euros (si le premier 6 apparaît seulement au 125</sup>e</sup> lancer), ..... un million d'euros (si le premier 6 apparaît seulement au millionième lancer) ... et on peut gagner beaucoup plus.

4.a. Expliquer très brièvement quel est le jeu auquel vous avez envie de participer
(le choix n'a pas besoin d'être réfléchi, il peut être intuitif). 


#### Réponse

4.b. Estimer à l'aide d'un programme python l'espérance de gain du jeu que vous avez choisi.

```python
from random import randint

def esperance_jeu_dix_des(n):
    gain_total = 0
    for i in range(n):
        gain = 0
        for i in range(10):  # Dix lancers de dé
            if randint(1, 6) == 6:
                gain += 6  # Gagne 6 euros pour chaque 6
        gain_total += gain
    return gain_total / n  # Gain moyen sur les simulations

def esperance_jeu_infini(n):
    gain_total = 0
    for i in range(n):
        gain = 0
        while True:
            gain += 1  # Gagne 1 euro par lancer
            if randint(1, 6) == 6:  # Arrête au premier 6
                break
        gain_total += gain
    return gain_total / n  # Gain moyen sur les simulations
```

```python
esperance_jeu_dix_des(100), esperance_jeu_infini(100)
```
