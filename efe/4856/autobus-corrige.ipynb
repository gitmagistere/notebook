Distribution des intervalles entre deux bus d'après une étude de la compagnie:

```python
import numpy as np
from numpy import random as rd

def intervalle():
  # Loi de probabilité des intervalles
  alea = rd.uniform()
  # On simule la loi grâce à la loi uniforme
  if (alea <= 0.1):
    return 2
  elif (0.1 < alea and alea <= 0.3):
    return 6
  elif (0.3 < alea and alea <= 0.8):
    return 10
  else:
    return 18
```

La fonction ci-dessous calcule l'intervalle moyen entre deux bus. Compléter ce programme

```python
def intervalle_moyen(N):
  S = 0
  for i in range(N):
    S = S + intervalle()
  return S/N
```
