```
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd   
# pandas pour analyse de donnÃ©es, ici pour la lecture du fichier de donnÃ©es csv

url = 'https://raw.githubusercontent.com/ela1687/donnees/main/aso.csv'
donnees = pd.read_csv(url)
# Importation et lecture des donnÃ©es depuis l'annÃ©e 1229

annees = []
S = 0
deltas = []
# durÃ©es sÃ©parant les Ã©ruptions

for d in donnees:
  annees.append(int(d)) 

for i in range(len(annees)-1):
  attente = annees[i]-annees[i+1]
  deltas.append(attente)
deltas.sort()
# on range les durÃ©es entre Ã©ruptions 

z = np.arange(0,50,5)
plt.hist(deltas, z,density=True ) 
plt.title("histogramme") 
plt.show()
# ReprÃ©sentation 
```
