$\pi$  est le périmètre d'un cercle $C$ de diamètre 1.

Vers 250 av. J.C., pour déterminer une valeur approchée de ce nombre, Archimède décida de considérer des polygones réguliers ayant le même nombre de côtés, l'un étant inscrit dans le cercle $C$ (en bleu) et l'autre lui étant circonscrit (rouge).

Polygones Archimede.png

On note pour tout $n\geq3$, $S_n$ le périmètre du polygone régulier à $n$ côtés inscrit dans le cercle $C$ et $T_n$ le périmètre du polygone régulier à $n$ côtés circonscrit à ce cercle.

On a alors pour tout $n\geq3$ :  $S_n\leq \pi\leq T_n$

# Formules de récurrence des suites $S_n$ et $T_n$

On peut démontrer que:

$S_n={n}\times{sin(\frac{\pi}{n})}$ et $T_n={n}\times{tan(\frac{\pi}{n})}$

et en déduire que :
$T_{2n}=\frac{2S_nT_n}{S_n+T_n}$ et $S_{2n}=\sqrt{S_n T_{2n}}$


# Ecrire un programme Python permettant de générer les suites $(T_{2^n})$ et $(S_{2^n})$

**1)** Calculer $T_4$ et $S_4$

Réponse :

$T_4=4$ et $S_4=2\sqrt2$

**2)** Compléter le programme de Python ci-dessous :

```python
from math import *

def archimede(n) :
	T = 4
	S = 2*sqrt(2)
	for k in range(n) :
		T = (2*S*T)/(S+T)
		S = sqrt(S*T)
	print("T =", T, "S =", S)
```

**3)** Tester son programme

```python
archimede(10)
archimede(50)
```
