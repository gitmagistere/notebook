Voici un exemple de programme permettant de tracer les courbe des fonctions carrée et cube sur [-3 ; 3].

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,3,100)
y=x**2
y2=x**3

plt.plot(x,y,'r', label = 'fonction carrée')
plt.plot(x,y2,'b', label = 'fonction cube')
plt.legend()
plt.grid()
plt.show()
```
