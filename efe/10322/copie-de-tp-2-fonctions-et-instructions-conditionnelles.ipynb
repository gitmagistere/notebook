<h2 style="text-decoration:underline" id="partie2">2- Instruction conditionnelle</h2>

Une instruction conditionnelle permet d'effectuer des choix dans un programme en fonction de la valeur vraie ou fausse d'une condition :

* si la condition est vraie, on effectue des instructions ;

* si la condition est fausse, on effectue d'autres instructions.

**Exemple 1** 

On suppose qu'un magasin de reprographie applique les tarifs suivants :

* 0,15 € l'unité jusqu'à 50 photocopies ;

* 0,10 € au-delà.

Voici le script d'une fonction <code>prix_photocopies</code> qui prend comme argument le nombre <code>n</code> de photocopies et qui renvoie le prix à payer :

```python
def prix_photocopies(n):
    if n<=50:
       prix=0.15*n
    else:
       prix=0.10*n
    return(prix)  
```

**Remarques**

* La condition <code>n<=50</code> correspond à "n inférieur ou égal à 50".
    
* Au niveau des lignes 3 et 5 il y a une indentation car les lignes précédentes se terminent par <code>:</code>. Le retour à l'indentation précédente aux lignes 4 et 6, qui ne se fait pas automatiquement, signale la fin du bloc d'instructions.
