````python
b = 5
print(f"La valeur de la variable b est {b}.")  # b est de type integer
````

````python
c = float(5)
print(f"La valeur de la variable c est {c}.")  # c est de type float
````

````python
a=4
b=5
print(f"Le produit de {a} par {b} est {a*b}") 
````

````python
a = 4
b = 5
print(f"{a}*{b}={a*b} et {a}/{b}={a/b}")
````

Voici  un premier exemple

````python
from turtle import *
t = Turtle()

t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
done()
````

On obtient un carr√© de 60 pixels de cot√© et le crayon se trouve au centre de la figure  au point de coordonn√©e (0,0).  


Que se passe-t-il ?

Une tortue marche avec un crayon. Elle trace ainsi ses d√©placements.

Entre les deux, on commande la tortue avec les instructions suivantes :

- **`forward(n)`**:  avance tout droit sur une distance de n pixels
- **`left(a)`**:  tourne la tortue d'un angle de a degr√©s vers la gauche
- **`right(a)`**:  tourne la tortue d'un angle de a degr√©s vers la droite.
- **`goto(x,y)` ou `setpos()`**:  D√©place la tortue vers le point de coordonn√©es (x,y)
- **`pencolor(str)`**:  d√©finit la couleur du crayon ('red', 'blue', 'green', etc.)
- **`pensize(n)`**: √©paisseur du crayon en pixel
- **`penup()`**:  l√®ve le crayon, la tortue se d√©place mais n'√©crit plus
- **`pendown()`**:  redescend le crayon, la tortue se d√©place en √©crivant de nouveau
- **`width(n)`**:  d√©finit l'√©paisseur-  du trait
- **`reset()`**:  pour r√©initialiser.
- **`circle(rayon, extent=None, steps=None)`**:  Dessine un cercle de rayon radius. 
- **`done()`** ou **`mainloop()`**:  D√©marre la boucle d'√©v√©nements et  permet de ne pas fermer la fen√™tre Turtle.

R√©p√©ter le code  (tracer un petit trait) deviens vite <b><i>Fastidieux !</i></b>

Quand une t√¢che peut √™tre effectu√©e plusieurs fois dans un programme, il devient vite TR√àS INT√âRESSANT d'utiliser un morceau de programme qu'on va r√©utiliser:  

- on isole alors ce morceau de code dans un <b><i>BLOC</i></b> qu'on appelle  <b><i>fonction</i></b>
- On pr√©cise son nom et en g√©n√©ral des param√®tres d'entr√©e et une valeur de retour.  

<div class="alert alert-block alert-warning" style="border-left:15px solid #EBA83A;">
    

En Python on d√©finit une fonction ainsi :
```python
def mafonction(parametre1, parametre2):
    #bloc d'instructions (optionnel)
    return valeur
```
    
    
</div>

Par exemple la fonction ***`carre(x)`*** qui prend en param√®tre un nombre $x$ et qui retourne son carr√©, s'√©crira :

````python
def carre(x):
    return x**2
````

Pour utiliser cette fonction ***carre(x)*** on √©crira:

````python
print(carre(3))
print(carre(4))
print(carre(5))
````

ou avec les **f-strings** ce qui est plus complet et joli.

````python
print(f"Le carr√© de 2 est {carre(2)}")
print(f"Le carr√© de 3 est {carre(3)}")
print(f"Le carr√© de 4 est {carre(4)}")
print(f"Le carr√© de 5 est {carre(5)}")
````

Une fonction peut prendre plusieurs param√®tres.  

En math√©matique on a la formule  $\overrightarrow{AB}\begin{pmatrix} x_B-x_A \\ y_B-y_A \end{pmatrix}$

La fonction ***coord_vecteur(xA, yA, xB, yB)*** qui prend en param√®tres  les coordonn√©es des points $A(x_A, y_A)$ et $B(x_B, y_B)$,   et qui renvoie les coordonn√©es du vecteur $\overrightarrow{AB}$ peut s'√©crire :

````python
def coord_vecteur(xA, yA, xB, yB):
    return (xB - xA, yB - yA)


print(coord_vecteur(1, -3, 4, 6))
````
