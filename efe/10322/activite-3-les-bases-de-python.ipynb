### 2.2 Afficher  (print)

Pour afficher en python on utilise la fonction  ***print()***

````python
print("Bonjour")
print("Ceci est du texte")
print(3)
print(2.5)
print("Le", "lycée", "pablo", "Neruda")
````

Pour afficher le contenu d'une variable dans du texte il existe plusieurs méthodes

#### Méthode 1: 

````python
a = 'cinq'
print("Le contenu de la variable a est", a, ".")
````

````python
b = 5
print("Le contenu de la variable b est", b, ".")
````

Cette méthode est parfois limité, il est alors préférable d'utiliser la deuxième méthode.

####  Méthode 2: en utilisant les  ***f-strings***

Je vous conseille cette méthode.

On met un **f** devant les guillemets et remplace ***{a}*** par le contenu de la variable ***a***. Peu importe si ***a*** est un texte, un flottant ou un entier,  le résultat sera affiché.

*Remarque:*  **f** signifie **format** (formater).

````python
a = "5"
print(f"La valeur de la variable a est {a}.")  # a est de type string
````

````python
b = 5
print(f"La valeur de la variable b est {b}.")  # b est de type integer
````

````python
c = float(5)
print(f"La valeur de la variable c est {c}.")  # c est de type float
````

````python
a=4
b=5
print(f"Le produit de {a} par {b} est {a*b}") 
````

````python
a = 4
b = 5
print(f"{a}*{b}={a*b} et {a}/{b}={a/b}")
````

Voici  un premier exemple

````python
from turtle import *
t = Turtle()

t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
done()
````

On obtient un carré de 60 pixels de coté et le crayon se trouve au centre de la figure  au point de coordonnée (0,0).  


Que se passe-t-il ?

Une tortue marche avec un crayon. Elle trace ainsi ses déplacements.  

<div class="alert alert-danger" style="border-left: 15px solid #a94442;">
    
On a écrit l'instruction **`from turtle import * `**  au début pour importer toutes les fonctions du module **« turtle »**.  
    
</div>

<div class="alert alert-danger" style="border-left: 15px solid #a94442;">
   La dernière instruction, <b><code>done()</code></b>, permet de demander l’affichage du tracé. Si on l’oublie, il ne se passe rien. 🤔
</div>



Entre les deux, on commande la tortue avec les instructions suivantes :

- **`forward(n)`**:  avance tout droit sur une distance de n pixels
- **`left(a)`**:  tourne la tortue d'un angle de a degrés vers la gauche
- **`right(a)`**:  tourne la tortue d'un angle de a degrés vers la droite.
- **`goto(x,y)` ou `setpos()`**:  Déplace la tortue vers le point de coordonnées (x,y)
- **`pencolor(str)`**:  définit la couleur du crayon ('red', 'blue', 'green', etc.)
- **`pensize(n)`**: épaisseur du crayon en pixel
- **`penup()`**:  lève le crayon, la tortue se déplace mais n'écrit plus
- **`pendown()`**:  redescend le crayon, la tortue se déplace en écrivant de nouveau
- **`width(n)`**:  définit l'épaisseur-  du trait
- **`reset()`**:  pour réinitialiser.
- **`circle(rayon, extent=None, steps=None)`**:  Dessine un cercle de rayon radius. 
- **`done()`** ou **`mainloop()`**:  Démarre la boucle d'événements et  permet de ne pas fermer la fenêtre Turtle.

## 4 Les fonctions

### 4.1 La syntaxe des fonctions


Répéter le code  (tracer un petit trait) deviens vite <b><i>Fastidieux !</i></b>

Quand une tâche peut être effectuée plusieurs fois dans un programme, il devient vite TRÈS INTÉRESSANT d'utiliser un morceau de programme qu'on va réutiliser:  

- on isole alors ce morceau de code dans un <b><i>BLOC</i></b> qu'on appelle  <b><i>fonction</i></b>
- On précise son nom et en général des paramètres d'entrée et une valeur de retour.  

<div class="alert alert-block alert-warning" style="border-left:15px solid #EBA83A;">
    

En Python on définit une fonction ainsi :
```python
def mafonction(parametre1, parametre2):
    #bloc d'instructions (optionnel)
    return valeur
```
    
    
</div>

Par exemple la fonction ***`carre(x)`*** qui prend en paramètre un nombre $x$ et qui retourne son carré, s'écrira :

````python
def carre(x):
    return x**2
````

Pour utiliser cette fonction ***carre(x)*** on écrira:

````python
print(carre(3))
print(carre(4))
print(carre(5))
````

ou avec les **f-strings** ce qui est plus complet et joli.

````python
print(f"Le carré de 2 est {carre(2)}")
print(f"Le carré de 3 est {carre(3)}")
print(f"Le carré de 4 est {carre(4)}")
print(f"Le carré de 5 est {carre(5)}")
````

Une fonction peut prendre plusieurs paramètres.  

En mathématique on a la formule  $\overrightarrow{AB}\begin{pmatrix} x_B-x_A \\ y_B-y_A \end{pmatrix}$

La fonction ***coord_vecteur(xA, yA, xB, yB)*** qui prend en paramètres  les coordonnées des points $A(x_A, y_A)$ et $B(x_B, y_B)$,   et qui renvoie les coordonnées du vecteur $\overrightarrow{AB}$ peut s'écrire :

````python
def coord_vecteur(xA, yA, xB, yB):
    return (xB - xA, yB - yA)


print(coord_vecteur(1, -3, 4, 6))
````

Remarque: (3,9) est un type particulier en python et s'appelle un ***tuple***.

Il existe aussi des fonctions sans paramètres d'entrée ou sans valeur de retour. Par exemple, une fonction qui affiche le nom complet d'un acronyme peut s'écrire :

````python
def nsi():
    print("Numérique et Sciences Informatiques")


nsi()
nsi()
````
