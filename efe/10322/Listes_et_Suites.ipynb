### Afficher les premiers termes d'une suite, notion de liste

Pour afficher automatiquement les 20 premiers termes de cette suite, on peut utiliser une *boucle* et la fonction <code>print</code>.

<div style="padding:1em ;border:solid 2px #AED6F1; border-radius:5px; background-color:#EBF5FB;">  
<div style="color:#5DADE2 ;">
    
## À FAIRE   
</div>
    
Exécuter la cellule suivante.

</div>

```python
for n in range(20):
    print(v(n))
```

L'inconvénient d'utiliser la fonction <code>print</code> est que les valeurs calculées sont certes affichées, mais sont ensuite perdues. Il est plus avantageux de créer une *liste* où l'on va stocker automatiquement les 20 premiers termes, pour pouvoir les réutiliser :

<div style="padding:1em ;border:solid 2px #AED6F1; border-radius:5px; background-color:#EBF5FB;">  
<div style="color:#5DADE2 ;">
    
## À FAIRE   
</div>
    
Exécuter la cellule suivante.

*Après exécution de la cellule, Python n'affiche rien. La liste <code>liste20</code> est maintenant créée et peut être utilisée.*
    
</div>

```python
liste20=[]               # Initialisation : on crée une liste vide appelée liste20
for n in range(20):      # On répète 20 fois
    liste20.append(v(n)) # On ajoute v(n) à la fin de liste20
```

<div style="padding:1em ;border:solid 2px #F5B7B1; border-radius:5px; background-color:#FDEDEC;">   
<div style="color:#EC7063;">
    
## À RETENIR
</div>
    
* les listes $python$ sont délimitées par <code>[</code> et <code>]</code> ;
* on commence par *initialiser* la liste, en créant une variable <code>liste20</code> à laquelle on affecte une liste *vide* ;
* l'instruction <code>liste20.append(truc)</code> permet d'ajouter en fin de liste l'élément *truc*.

    
</div>

### Représentation graphique des premiers termes d'une suite

Le script suivant permet d'afficher les 20 premiers termes de la suite $(v_n)$.

<div style="padding:1em ;border:solid 2px #AED6F1; border-radius:5px; background-color:#EBF5FB;">  
<div style="color:#5DADE2 ;">
    
## À FAIRE   
</div>
    
Exécuter le script suivant.

*L'exécution est un peu longue, c'est normal.*
    
</div>

```python
import matplotlib.pyplot as plt


N = 20              # On veut afficher 20 termes

X = []              # On crée une liste vide qui contiendra les abscisses des points à représenter
Y = []              # On crée une liste vide qui contiendra les ordonnées des points à représenter

for i in range(N):  # On répète 20 fois
    X.append(i)     # On ajoute l'entier i à la liste des abscisses
    Y.append(v(i))  # On ajoute v(i) à la liste des ordonnées
    
plt.clf()           # On efface (si besoin) le graphique précédent
plt.plot(X,Y,"bo")  # On place les points qui ont pour abscisses les termes de la liste X et pour ordonnées ceux de la liste Y
plt.show()
plt.close()
```
