<p style="text-align:justify";>Dans un programme, on a souvent besoin de répéter plusieurs fois la même instruction (ou séquence d'instructions).
On suppose dans ce chapitre que l'on sait à l'avance combien de fois on va devoir répéter la séquence d'instructions. La <strong>boucle <code>for</code></strong> (ou boucle bornée) permet de répondre à ce besoin.<br>
    <u>Remarque</u> : Si on ne sait pas à l'avance combien de fois on va répéter la boucle (par exemple si on doit arrêter la boucle lorsqu'une certaine condition est réalisée), on utilisera la boucle <code>while</code> (tant que) qui sera vue dans le chapitre 5.</p>

<h2 style="text-decoration:underline;" id="partie1">1- Boucle bornée (boucle <code>for</code>)</h2>

Le programme ci-dessous est un exemple simple d'utilisation de la boucle <code>for</code>.<br>
On répète ici 5 fois l'instruction <code>print("Bonjour")</code>.<br>
Le bloc d'instructions à répéter doit être **indenté**, c'est-à-dire décalé de quelques espaces vers la droite.<br>
La variable <code>i</code> est l'indice de boucle (ou variable de boucle ou compteur de boucle). On peut lui donner le nom que l'on veut. Son utilisation sera vue dans la partie <em><a href="#partie2">2- Utilisation du compteur de boucle</a></em>.

```python
for i in range(5):
    print("Bonjour")
```

**Boucles imbriquées**

<p style="text-align:justify";>On a parfois besoin d'insérer une boucle à l'intérieur d'une autre boucle, par exemple lorsqu'on souhaite parcourir successivement tous les pixels d'une image. On parcourt l'image ligne par ligne, et on parcourt chaque ligne colonne par colonne. <br>
La structure est la suivante :</p>

```python
for i ...
    for j ...
        instructions
        instructions        
```

<h2 style="text-decoration:underline;" id="partie4"> 2 - Dessiner avec Turtle</h2>

La bibliothèque **turtle** permet de dessiner à l'écran.<br>
Le petit programme commenté ci-dessous permet d'obtenir un tracé intéressant :

```python
from turtle import *      # On importe la bibliothèque turtle

speed(10)                 # On règle la vitesse du tracé (un entier compris entre 1 et 10)
color('red', 'yellow')    # On fixe la couleur du tracé à 'rouge' et la couleur de rempissage à 'jaune'
begin_fill()
for i in range(36):
    forward(200)          # La tortue avance de 200 pixels 
    left(170)             # La tortue tourne vers la gauche de 170°
end_fill()
done()                    # Il faut terminer par done() pour lancer l'exécution du tracé    
```
