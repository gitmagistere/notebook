```python
"""
Programme avec boucle For
"""
nbr_nucleotide = 0
dist_bases = 0.34
for k in range(22):
    fichier = "D:/chr"+str(k+1)+".fna"
    chromo = open(fichier)
    contenu = chromo.read()
    nbr_nucleotide = nbr_nucleotide + len(contenu)
lg_nbr_nucleotide = nbr_nucleotide * dist_bases

print(round(nbr_nucleotide,2))
    
```
