# Les instructions conditionnelles

Une instruction conditionnelle permet à un programme de réagir différemment en fonction de la valeur d'une expression booléenne. En python, on utilise le mot clé if que l'on peut compléter avec les mots clés optionnels elif et else. On peut mettre autant de elif que nécessaire mais il y a au plus un if et un else.


Cela s'apparente à dire à l'ordinateur :
~~~~~~
si expression:
	suite_instruction1
sinon:
	suite_instruct2
suite_instruction3

~~~~~~




Si "expression" est **<span style="color: green">True</span>**
(vraie) alors "suite_instruction1" est exécuté et "suite_instruction2" est ignoré.

Sinon (sous-entendu que "expression" est **<span style="color: green">False</span>**
(fausse)) "suite_instruction2" est exécuté et "suite_instruction1" est ignoré.


Notez l'indentation «suite_instruction1» et de «suite_instruction2». Notez également les ":" qui sont nécessaire à la syntaxe de Python.

Dans tous les cas, « suite_instruction3 » sera exécuté.

Si plusieurs conditions doivent être testées à la suite, on peut utiliser la suite d’instructions : 
~~~~~
si condition2:
    suite_instructions1
sinonsi condition2:
    suite_instructions2
sinon suite_instruction3
~~~~~

En Python :
~~~ python
if condition1:    
    instructions 
elif condition2:    
    instructions2 
else:
    instructions3
~~~


### Exemple 
Avant d'exécuter la cellule, essayez de déterminer l'affichage attendu.

```python
nombre_a = 12
nombre_b = 13
if nombre_a < nombre_b:
    print(f"{nombre_a} est plus petit que {nombre_b}")
else:
    print(f"{nombre_b} est plus petit que {nombre_a} ou égal à {nombre_a}")
```
