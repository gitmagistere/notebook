# Exemples

```python
def seuil():
    aire = 3     # aire initiale
    jours = 0
    while aire <= 90:   # tant que le test logique est vrai ...
        aire = aire * 2
        jours = jours + 1
    return jours, aire
```

```python
def temps(C0, taux):
    capital = C0
    annees = 0
    while capital < 2 * C0:  # tant que C0 n'est pas au moins doublé
        capital = capital * (1 + taux)
        annees = annees + 1
    return annees, capital
```

```python
def syracuse(n):
    sequence = [n]      # liste des étapes
    while n > 1:         # tant que 1 n'est pas atteint ...
        if n % 2 == 0:   # si n est pair
            n = n//2     
            sequence.append(n)  # on ajoute n à la séquence
        else:           
            n = 3*n+1
            sequence.append(n)
    return sequence     
```
