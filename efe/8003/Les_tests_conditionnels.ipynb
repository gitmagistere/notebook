# Exemple 1
Dans le programme ci-dessous, la fonction recherche(mot) renvoie l'index du mot
recherché si ce mot est inclus dans la chaîne de caractères nommée brin.

```python
brin = "ATGACATTCAGGTATACTCCGTATTAAGCAACGTAGCTGAATATGCTAAGTGATTACATAG"

def recherche(mot):
    if mot in brin:  # si le mot est inclus dans la chaîne brin. 
        return brin.index(mot)
```

# Exemple 3

Un cinéma propose deux formules tarifaires :

★ Tarif A : un abonnement de 30 euros puis, un prix de 4 euros par séance.

★ Tarif B : sans abonnement avec un prix de 6 euros par séance.

```python
def tarif(n):
    tarifA = 4*n + 30
    tarifB = 6*n
    if tarifA == tarifB:
          return 'A et B sont equivalents'
    elif tarifA < tarifB:
         return 'A est plus avantageux'
    else:
         return 'B est plus avantageux'
```

Exécuter le programme ci-dessus, puis, appeler tarif(5), tarif(15) puis, tarif(20) dans la console ci-après.
