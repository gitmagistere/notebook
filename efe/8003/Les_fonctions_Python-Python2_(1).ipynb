# Exemples

```python
from random import randint
# dans la librairie random, randint permet 
# de générer des entiers pseudo-aléatoires

def somme():       
  # on remarque que cette fonction ne prend aucun paramètre
  d1 = randint(1,6)  # entier pseudo-aléatoire entre 1 et 6
  d2 = randint(1,6)
  return d1 + d2
```

```python
def farenheit(Tc):    # la fonction dépend du paramètre Tc
    Tf = 1.8*Tc + 32  # conversion en degrés Farenheit
    return Tf         # renvoie la température en Farenheit
```

```python
def moyenne(a, b, c):    # 3 paramètres a, b et c
    m = (a + b + c)/3  
    return m      
```

```python
def carre(x):      
    return x**2    # x puissance 2 

def cube(x):      
    return x**3    # x puissance 3

def comparaison(x):
    return cube(x) <= carre(x)
```
