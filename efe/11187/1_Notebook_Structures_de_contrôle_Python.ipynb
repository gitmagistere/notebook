Comme indiqué ci-dessous :

Si la condition est vérifiée, alors les instructions 1 sont effectuées sinon ce sont les instructions 2 qui le sont.\
Remarques : <li>On peut utiliser le if sans le else
<li>On peut construire des conditions élaborées avec des 'and' ou des 'or'

```python
# que répond le programme ci dessous ?

a=100%7

if (a%2==0):
    print(a,' est pair')
else :
    print(a,' est impair')
```

<h1> Les boucles "for" <h1>

On utilise une boucle 'for' quand on veut répéter des instructions un nombre de fois déterminé à l'avance. 
La structure d'une boucle for est la suivante :



    
i est un compteur et les instructions 1 et 2 seront répétées pour les valeurs du compteur **allant de p jusqu'à n-1**. Si le nombre p n'est pas précisé, le compteur commence d'office à 0.
    
 Pour les programmes suivants, combien de fois sera répété l'instruction print ?   

```python
for i in range(0,5):
    print('bonjour')
```

```python
for i in range(4,7):
    print('hello')
```

```python
for i in range(3):
    print('salut')
    
```
