Création de la fonction `miroir` permettant d'obtenir la liste "miroir" d'une liste donnée
 </div>





```python
def miroir(L):
    m = [0 for i in range(len(L))]
    for k in range(len(L)):
        if L[k] == 0 :
            m[len(L) - 1 - k] = 1
        else :
            m[len(L) - 1 - k] = 0 # inutile car m est une liste composée initialement uniquement de 0
    return m

from tutor import * # permet d'appeler le module PythonTutor

miroir([0,0,1,0,0,1,1])
tutor()
```

Création la fonction `parcours` renvoyant la liste permettant de parcourir le dragon (suite de 0 et de 1)

**Le 0 correspond à Droite et le 1 correspond à Gauche**

```python
def parcours(n):
    L = [0]
    for i in range(1,n):
        G = miroir(L)
        L.append(0)
        L = L + G
    return L

parcours(4)
```

Création d'une fonction permettant de construire la courbe du dragon

```python
from turtle import *
color('red')
def trace(pas,n):
    for x in parcours(n):
        forward(pas)
        if x == 0 :
            right(90)
        else :
            left(90)
trace(20,4)    
done()        
```
