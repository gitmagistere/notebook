## Déterminer un encadrement de $\sqrt{2}$ d'amplitude $10^{-n}$

```
 def BalayageRacine2(pas):
    x=1
    while x*x <=2:
        x = x + pas
    return (x-pas,x)
```

## Déterminer si un entier nautrel $a$ est multiple d'un entier natruel $b$

```
 def AestmultipledeB(a,b):
    if a%b == 0:
        return True
    else:
        return False
```

```
AestmultipledeB(4054,4)
```

```
def AestmultipledeB_2(a,b):
    n = 0
    while n*b < a:
        n = n+1
    if n*b == a:
        return True
    else:
        return False
```

```
AestmultipledeB_2(4056,4)
```

## Pour des entiers $a$ et $b$ donnés, déterminer le plus grand multiple de $b$ inférieur ou égal à $a$

```
def PlusGrandMultiple(a,b):
    n = 1
    while n*b <= a:
        n += 1
    return n*b - b
```

```
PlusGrandMultiple(172, 17)
```

## Déterminer si un entier naturel est premier

```
def EstPremier(n):
    for i in range(2, n):
        if n%i == 0:
            return False
    return True
```

## Déterminer la première puissance d'un nombre positif donné supérieure ou inférieure à une valeur donnée

```
def PremièrePuissance(a,b):
    p = 1
    while p < a:
        p *= b
    return p/b
```

## Étudier l'alignement de trois points du plan

### En calculant les équations des droites

```
def EquationDroite(A,B):
    a = (B[1]-A[1])/(B[0]-A[0])
    b = A[1] - a*A[0]
    return a,b
```

```
def Alignés(A,B,C):
    if (abs(B[0]-A[0])<10**(-12) and abs(B[1]-A[1])<10**(-12)):
        return True
    else:
        a,b = EquationDroite(A,B)
        if abs(C[1]-a*C[0]-b)<10**(-12):
            return True
        else:
            return False
```

### En calculant le déterminant des vecteurs

```
def déterminant(v1,v2):
    return v1[0]*v2[1]-v1[1]*v2[0]
```

```
def Alignés_2(A,B,C):
    v1,v2 = (B[0]-A[0],B[1]-A[1]),(C[0]-A[0],C[1]-A[1])
    return True if abs(déterminant(v1,v2)) < 10**(-12) else False
```

## Pour une fonction dont le tableau de variations est donné, algorithmes d'approximation numérique d'un extrêmum(balayage, dichotomie)

```
def f(x):
    return x**3 + 3*x**2-2*x+1
```

```
def BalayerExtreme(f,a,b,pas):
    x = a
    max = f(x)
    while x <= b:
        x += pas
        if f(x)>max:
            max = f(x)
            X = x
    if X > b:
        return b,f(b)
    else:
        return X,max
```

```
BalayerExtreme(f,-4,0,0.001)
```

## Calcul approché de la longueur d'une portion de courbe représentative

```
from math import sqrt
```

```
def distance(A,B):
    return sqrt((B[1]-A[1])**2+(B[0]-A[0])**2)
```

```
def f(x):
    return x**3 + 3*x**2 -2*x +1
```

```
def LongueurCourbe(f,a,b,pas):
    x = a
    L = 0
    while x < b:
        A = (x,f(x))
        B = (x+pas,f(x+pas))
        L = L + distance(A,B)
        x += pas
    return L
```

```
LongueurCourbe(f,-2,2,0.001)
```

## Pour des données réelles ou issues d'une simulation, lire et comprendre une fonction écrite en Python renvoyant la moyenne $m$, l'écart-type $s$ et la proportion d'éléments appartenant à l'intervalle $[m-2s;m+2s]$

```
import numpy as np
```

```
def Paramètres(Liste):
    m = np.mean(Liste)
    s = np.std(Liste)
    n = len(Liste)
    effectif = 0
    for k in Liste:
        if (k > m-2*s) and (k < m+2*s):
            effectif += 1
    return m,s,effectif/n
```

La fonction suivante génère une liste de `n` observations d'une variable aléatoire suivant une loi normale de moyenne `mu` et d'écart-type `sigma`. 

```
def Simulation(n,mu,sigma):
    return sigma*np.random.randn(n)+mu
```

```
L = Simulation(1000, 10, 1)
print(L)
```

```
Paramètres(L)
```

La fonction suivante permet de générer une représentation graphique de la série. 

```
import matplotlib.pyplot as plt
```

```
def Graphique(Liste):
    n = len(Liste)
    X = range(1,n+1)
    m,s,f=Paramètres(Liste)
    plt.figure(figsize=(20,10))
    plt.plot(X,Liste,'b+')
    plt.grid()
    plt.axhline(y=m, lw=2, color='g')
    plt.axhline(y=m+2*s, lw=2, color='r')
    plt.axhline(y=m-2*s, lw=2, color='r')
    return None
```

## Lire et comprendre une fonction Python renvoyant le nombre ou la fréquence de succès dans un échantillon de taille `n` pour une expérience aléatoire à deux issues

```
import random
```

### Avec une fonction intégrée

```
def Succès(n,p):
    tirage = random.choices([1,0],weights=[p,1-p],k=n)
    s = tirage.count(1)
    return s, s/n
```

```
Succès(1000,0.4)
```

### Sans fonction intégrée

```
def Succès_2(n,p):
    s = 0
    for k in range(n):
        if random.random() < p:
            s += 1
    return s, s/n
```

```
Succès_2(1000,0.4)
```

```
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from random import *
from IPython.display import HTML

###########
N = 300 # Nombre d'expériences aléatoires à réaliser 
###########

def de_truque():
    DE = [1,1,2,3,4,5]
    return choice(DE)

def frequences(n):
    Liste_Y=[0]
    nb_succes = 0
    for compteur in range(1,n+1):
        if de_truque()==1:
            nb_succes = nb_succes + 1
        Liste_Y = Liste_Y + [nb_succes / compteur]
    return Liste_Y

fig, ax = plt.subplots(figsize=(10,5))
xdata, ydata = [], []
points, = plt.plot([], [], 'b+')
Liste_a_representer = frequences(N)

def init():
    ax.set_xlim(0, N)
    ax.set_ylim(0, 1)
    return points,

def update(i):
    xdata.append(i)
    ydata.append(Liste_a_representer[i])
    points.set_data(xdata, ydata)
    return points,

ani = FuncAnimation(fig, update, frames=N, init_func=init, blit=True)
HTML(ani.to_jshtml())
#HTML(ani.to_html5_video())
```

## Simuler N échantillons de taille n d’une expérience aléatoire à deux issues. Si $p$ est la probabilité d’une issue et $f$ sa fréquence observée dans un échantillon, calculer la proportion des cas où l’écart entre $p$ et $f$ est inférieur ou égal à $\dfrac{1}{\sqrt{n}}$

```
import random
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
```

```
def Echantillon(n,p):
    tirage = random.choices([1,0], weights=[p,1-p], k=n)
    e = tirage.count(1)
    f = e/n
    return tirage, e, f
```

```
def Proportion(N,n,p):
    frequences = np.array([Echantillon(n,p)[2] for k in range(N)])
    F = np.sum(abs(frequences-p)<=1/sqrt(n))
    F /= N
    X = range(1,N+1)
    plt.figure(figsize=(20,10))
    plt.plot(X,frequences,'rx')
    plt.grid()
    plt.axhline(y=p,color='b', lw=3)
    plt.axhline(y=p-1/sqrt(n), color='g')
    plt.axhline(y=p+1/sqrt(n), color = 'g')
    plt.title(f"Proportion des fréquences dans l'intervalle : {F}")
    return F
```

```
Proportion(1000,100,0.7)
```
