On veut modifier le programme préccédent pour qu'à un nombre entier `n` (supérieur ou égal à 1) il renvoie la somme des entiers consécutifs de 1 à `n`.


```python
def somme(n) :
  s=0
  for i in range (1,n+1) :
    s=s+i
  return(s)
```

Calculer la somme des entiers de 1 à 100.

```python
somme(100)
```
