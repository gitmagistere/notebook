<h2 style="background-color: rgb(0,161,153); color: white; border-radius: 3px;padding: 10px;">✏️ Présentation de Basthon</h2>

L'utilisation d'un notebook permet, dans un même document, d'exécuter du code python, d'écrire du texte, d'insérer différentes ressources multimédia, d'utiliser le langage html, etc , mais sachez que la plupart du temps on utilise un *éditeur de code* pour écrire et exécuter des programmes Python. 

C’est un logiciel qui se décompose généralement en deux fenêtres : 

- la **fenêtre d’édition** (à gauche dans la capture d'écran ci-dessous) dans laquelle on écrit le programme dans le langage Python 
- et la **console** (à droite dans la capture d'écran) dans laquelle peuvent notamment s’afficher les valeurs des variables souhaitées.




Il existe énormément d'éditeur et nous utiliserons pour cette séance un éditeur en ligne appelé **BASTHON** (qui signifie **B**ac **À** **S**able Py**thon**) qui ne nécessite pas d’installation (vous pourrez donc l’utiliser chez vous rien qu’avec une connexion Internet). On peut également installer une version offline.
