## <div style = "color:#124660"> 3. Ajouter des tableaux en Markdown   &nbsp;$^{★★★}$</div>

Dans certains cas, il peut être intéressant de mettre des informations sous la forme d'un tableau.
Ce dernier nécessite une syntaxe d'écriture bien défini :   

\| En-tête 1 | En-tête 2| En-tête 3|   $\quad$*La 1<SUP>ère</SUP> ligne correspond aux en-têtes et chaque colonne est séparée par le caractère '|'*   
\| :------------ | :------------: | ------------: |   $\quad$*La 2<SUP>ème</SUP> ligne spécifie l'alignement en ajoutant ' : ' à gauche et/ou à droite avec des tirets*   
\|  $\;\;$ val 1 $\;\;$ |  $\;\;$ val 2 $\;\;$ |  $\;\;$ val 3 $\;\;$|   $\quad$*On ajoute les données colonnes par colonne*   
\|  $\;\;$ val 4 $\;\;$ |  $\;\;$ val 5 $\;\;$ |  $\;\;$ val 6 $\;\;$|   $\quad$*On ajoute autant de lignes qu'on le souhaite*   



Ce tableau en Markdown apparaîtra comme cela :   
| En-tête 1| En-tête 2 | En-tête 3 |   
| :------------- | :------------: | -------------: | 
|  val 1 |  val 2 |  val 3|
|  val 4 |  val 5 |  val 6|
