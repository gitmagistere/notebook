```python
import matplotlib.pyplot as plt




omega = np.linspace(0.95,1.05,2)
tau = np.linspace(-20,20,300)
Ntau = tau.size
Nomega = omega.size

I = np.zeros(Ntau)
for k in range(Ntau):
    I[k] = 2
    for i in range(Nomega):
        I[k] += 2*np.cos(2*np.pi*omega[i]*tau[k])/Nomega

print(omega, Nomega)
        
plt.figure()
plt.plot(tau,I)
plt.xlabel("tau")
plt.ylabel("Interference")
plt.axis([tau.min(),tau.max(),0,5])
```

```python
def psi(n,p):
    x = (np.random.sample(n)-0.5)*2*np.pi
    y = np.repeat(x,p)
    return y             
             



n=50000 # nombre de periodes
p=30 # temps de coherence
psi1 = psi(n,p)
psi2 = psi(n,p)

plt.figure(figsize=(12,6))
plt.plot(psi1,'b')
plt.plot(psi2,'r')
plt.ylabel('psi')
plt.xlabel('t')
plt.axis([0,3000,-np.pi,np.pi])
plt.grid()
             
```

```python
def interference(tau,psi1,psi2):
    psi = np.roll(psi1,int(tau))-psi2
    cos = np.cos(2*np.pi*tau+psi)
    return np.mean(cos)
             

tau = np.linspace(0,20,300)
N = tau.size
I = np.zeros(N)
for k in range(N):
    I[k] = interference(tau[k],psi1,psi2)
    
plt.figure()
plt.plot(tau,I)
plt.xlabel("tau")
plt.ylabel("Interference")
plt.axis([0,tau.max(),-1,1])

             
```
