```python
import numpy as np
import matplotlib.pyplot as plt

phi = np.linspace(-10,10,300)
Nphi = phi.size

racine = np.zeros(Nphi)
I = np.zeros(Nphi)
racine = 1+ 2*np.cos(phi)
I = racine * racine

        
plt.figure()
plt.plot(phi,racine)
plt.plot(phi,I)
plt.xlabel("phi")
plt.ylabel("Interference")
plt.axis([phi.min(),phi.max(),-2,10])
```

```python
N = 20
phi = np.linspace(0.0005,20,300)
Nphi = phi.size

racine = np.zeros(Nphi)
I = np.zeros(Nphi)
racine = np.sin(N*phi/2)/np.sin(phi/2)
I = racine * racine

        
plt.figure()
plt.plot(phi,I)
plt.xlabel("phi")
plt.ylabel("Interference")
plt.axis([phi.min(),phi.max(),0,N*N])
```
