# Suite geométrique:
Une suite géométrique est une suite de nombres définie par un premier terme $U_0$ ou $U_1$ (selon ce qui est "pratique") et une raison $q$


$$
\left( U_n \right) = \left\{
    \begin{array}{ll}
        U_1 \\
        U_{n+1}=q \times U_{n}
    \end{array}
\right.
$$

$$
\begin{align*}

\end{align*}
$$

## Sens de variation d'une suite géométrique
De manière générale:
   * Une suite $\left( U_n \right)$ est strictement croissante si:

$$U_{n+1}>U_n$$
   * Une suite $\left( U_n \right)$ est strictement décroissante si:

$$U_{n+1}<U_n$$

Une **suite géométrique** de premier terme strictement positif est:
   * croissante si $q>1$
   * décroissante si $q <1$
