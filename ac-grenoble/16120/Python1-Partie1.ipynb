## Autour de la division euclidienne

#### Définition

En programmation, on travaille la plupart du temps avec des entiers. On ne fait pas de division, mais une division euclidienne : la division euclidienne de $a$ par $b$ revient à trouver les réels $q$ et $r$ tels que l'on ait<br/>
$$a = b \times q +r$$
$$0 \leq r < b$$ <br/>
Le nombre $q$, appelé quotient, est donné par la commande Python : a//b.<br/>
Le nombre $r$, appelé reste, est donné par la commande Python : a%b.<br/>
Par exemple, pour obtenir le quotient et le reste dans la division euclidienne de 25 par 4 : 

````python
25//4
````

````python
25%4
````

La division euclidienne permet de répondre à la question : dans 100, combien de fois rentre-t-il 6 ?

````python
100//6
````

et alors, combien il reste ?

````python
100%6
````

C'est une fonctionnalité qui est très utile dans de très nombreux programmes.

## L'affectation `=`

L'affectation dans une variable se fait simplement avec le signe `=`. Le nom de la variable peut contenir des lettres en minuscule, majuscule, des chiffres (mais il ne peut pas commencer par un chiffre) et le caractère `_`.  
Nous vous recommandons d'utiliser des noms qui ont du sens, même s'ils sont très longs : vous éviterez ainsi les mots anglais du langage Python.  
L'affectation n'est pas un affichage : le fait d'affecter ne génére pas d'affichage.

````python
annee_courante = 2022 # bonne année !
````

````python
exemple1 = 7**2
````

````python
je_veux_faire_un_nom_de_variable_très_long = 12//3
````

La lecture se fait de "droite à gauche" : 
- j'affecte le nombre 2022 à la variable `annee_courante`
- j'affecte le résultat de $7^2$ à la variable `exemple1`
- j'affecte le résultat du quotient de la division euclidienne de 12 par 3 à la variable `je_veux_faire_un_nom_de_variable_très_long`.

Il faut donc bien avoir en tête que l'on réalise déjà le calcul qui est à droite, puis qu'on affecte ce résultat dans le nom de variable présent à gauche. Cela signifie que l'on peut alors **écraser** le contenu d'une variable par un autre.

````python
annee_suivante = annee_courante + 1
print(annee_suivante)
annee_suivante = annee_suivante + 1
print(annee_suivante)
````

## La fonction `print()`

Python est un langage de programmation international. Les noms de ces fonctions sont donc des noms anglais. `print()` permet de faire l'affichage d'une variable :

````python
print(annee_courante)
````

La fonction `print` peut prendre plusieurs paramètres : 

````python
print(annee_courante, exemple1)
````

La fonction `print` peut combiner du texte et des variables.  
En Python, le caractère `#` permet d'écrire du commentaire sur le code.

````python
# 1ere méthode
print("Nous sommes dans l'année",annee_courante)
# 2eme méthode
print(f"C'est l'année {annee_courante} !")
````

La méthode utilisée actuellement est la 2ème, appelée `f-string`. C'est celle que nous essaierons d'utiliser par la suite.

**Un premier exemple d'erreur :**

````python
print(b)
````

Ce bloc nous affiche une erreur. Lors de l'affichage de l'erreur, il y a deux choses qui vont nous intéresser : 
- la ligne à laquelle l'erreur est apparue : ici, c'est la ligne 1 (`line 1`) ;
- le nom de l'erreur : `NameError : name 'b' is not defined` 

Là encore, on est sur de l'anglais, mais cela reste de **l'anglais abordable** : le nom de variable `b` n'est pas défini. Donc Python ne sait pas quoi faire avec `b`...
