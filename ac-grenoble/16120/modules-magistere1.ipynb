# Les modules Python

Python fonctionne avec des bibliothèques - librairies - modules (vous trouverez les trois dénominations) : ce sont des ensembles de fonctions qui sont déjà programmées et que tout le monde peut utiliser. Elles sont toutes documentées. En mathématique en classe de seconde nous utiliserons principalement : 
- math : pour l'utilisation de la racine carré, le nombre pi, les fonctions sinus, cosinus, ...
- random : pour travailler sur des nombres aléatoires et faires des expériences aléatoires 
- matplotlib et numpy : pour tracer des fonctions. **Le code sera alors toujours donné car possédant trop de paramètres**.

# Le module `math` : 

Pour utiliser le module `math`, il faut saisir l'instruction `from math import *` en début de programme

````python
from math import *
sqrt(25)
````

# Le module `random` : 

Cette bibliothèque est relativement riche. Nous allons regarder quelques unes de ces fonctions.

Pour utiliser le module `random`, il faut saisir l'instruction `from random import *` en début de programme

````python
from random import *

print(f"Un nombre aléatoire dans l'intervalle [0;1[ est {random()}")
print(f"Un nombre aléatoire dans l'intervalle [5;12] est {randint(5,12)}")
print(f"Un nombre aléatoire dans l'intervalle [5;15[ est {randrange(5,15)}")
print(f"Un nombre aléatoire dans l'intervalle [0;100[ est {randrange(100)}")
````

# Le module `turtle` :

Le module `turtle` permet de faire des dessins. Il comporte de nombreux paramètres que vous pouvez retrouver sur [cette page](https://docs.python.org/fr/3/library/turtle.html#turtle-methods). 
- Pour afficher le dessin **dans Capytal**, il est nécessaire que le code se termine par **`done()`**.

````python
from turtle import *
forward(150)
right(90)
forward(150)
right(90)
forward(150)
right(90)
forward(150)
right(90)
done()
````

# Les modules `matplotlib` et `numpy` :

Ces modules permettent la représentation graphique de fonction. Le code pour obtenir une *belle* représentation graphique est cependant complexe. Nous allons donc simplement le donner pour l'instant. L'import de ces deux bibliothèques est relativement long.

````python
from matplotlib.pyplot import *
from numpy import arange
abscisses = [i for i in arange(-4,4.1,0.1)] # ensemble de définition de la fonction : ici [-4;4] et un pas de 0.1
images = [i**2 for i in abscisses] # expression algébrique de la fonction : ici, la fonction carré
plot(abscisses,images)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
show()
close()
````

## Bilan sur l'import d'une bibliothèque : 

L'appel à une bibliothèque se fait par le mot clé `from` suivi du nom de la librairie puis `import *`

Pour plus de rapidité, il est possible de n'importer que la fonction que l'on souhaite utiliser en remplacement de `*`  
- Exemple : `from math import sqrt` pour importer la racine carrée.
