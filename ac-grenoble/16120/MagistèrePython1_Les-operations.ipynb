### Autour de la division euclidienne

En programmation, on travaille la plupart du temps avec des entiers. On ne fait pas de division, mais une division euclidienne : la division euclidienne de $a$ par $b$ revient à trouver les réels $q$ et $r$ tels que l'on ait<br/>
$$a = b \times q +r$$
$$0 \leq r < b$$ <br/>
Le nombre $q$, appelé quotient, est donné par la commande Python : a//b.<br/>
Le nombre $r$, appelé reste, est donné par la commande Python : a%b.<br/>
Par exemple, pour obtenir le quotient et le reste dans la division euclidienne de 25 par 4 : 

````python
25//4
````

````python
25%4
````

#### Application à la parité d'un nombre

La division euclidienne permet aussi de savoir si un nombre est pair ou impair : il suffit de regarder si le reste dans la division euclidienne par 2 est 0 (nombre pair) ou 1 (nombre impair).  
C'est une fonctionnalité qui est très utile dans de très nombreux programmes.

Vérifions que 13 est bien impair :

````python
13%2
````
