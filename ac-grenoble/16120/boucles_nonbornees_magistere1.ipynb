### Les boucles non-bornées

Malheureusement, la plupart du temps, on ne sait pas combien de fois on va devoir recommencer une opération... Par exemple, compter le nombre de chiffre dans $2022^{2022}$ : si on avait déjà la réponse, il n'y aurait pas besoin de compter !   
Il nous faut donc une structure `tant que` qui nous permet de continuer à répéter une suite d'instruction jusqu'à ce qu'une certaine condition **ne se réalise plus**. <br/>
Il faudra donc faire attention à être sûr que la condition puisse à un moment ne plus se réaliser, sinon, on va créer une boucle infinie : le programme va continuer à fonctionner sans jamais s'arrêter. 

````python
from tutor import tutor
def nbre_chiffres(n):
    "Détermine le nombre de chiffres de l'entier n"
    i = 0
    while n // (10**i) !=0:
        i=i+1
    return i

print(nbre_chiffres(173897))
tutor()
````
