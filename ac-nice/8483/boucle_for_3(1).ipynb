Maintenant nous y sommes, on peut passer à la production "opérationnelle".

Ecrivez un programme qui va tirer des nombres de façon aléatoire compris entre 1 et 30. Ce programme doit être capable de demander à l'opérateur le nombre de tirages qu'il doit effectuer (donc il va y avoir une interaction).

Pour mémo on peut récupérer une entrée clavier de façon assez simple en utilisant input()...

Exemple:

```python
reponse=input("votre réponse ici même:")
print(reponse)
```

Par contre cette méthode a un petit défaut:

essayez de rentrer un nombre et vérifiez si le calcul proposé a bien lieu dans le programme suivant:

```python
print("entrez un nombre entre 1 et 30")
reponse=input("votre réponse ici même:")
resultat= 2*reponse
print(resultat)
```

Il faut donc bien imposer que ce qui va être saisi soit effectivement considéré par le programme comme un nombre et non une chaine de caractère !

On va donc imposer ce fait en formattant le contenu de la variable qui récupère l'entrée:

```python
print("entrez un nombre entre 1 et 30")
reponse=int(input("votre réponse ici même:")) #int permet d'imposer le fait que ce qui est entré est un nombre entier
resultat= 2*reponse
print(resultat)
```
