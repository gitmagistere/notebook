Nous allons voir un premier exemple simple d'utilisation de while

```python
i = 0         # on définit l'index témoin
while i<10 :  # on initie la boucle, vous remarquerez les :
  print(i**2) # on exécute l'opération si la condition n'est pas réalisée
  i = i+1     # on incrémente l'index
print('fin de boucle')
```

Vous remarquerez que le résultat 100 ne s'affichera pas puisque 10 est supérieur à la condition de vérification
