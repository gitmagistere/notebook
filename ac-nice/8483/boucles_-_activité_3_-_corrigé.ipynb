```python
from random import randint #on importe le module "randint" depuis la bibliothèque "random"

print("Sur quel intervalle souhaitez vous travailler ?")
inf=int(input("borne inférieure : ")) #on définit la borne inférieure de l'échantillon de tirage
sup=int(input("borne supérieure : ")) #on définit la borne supérieur de l'échantillon de tirage
nb_tirage=int(input("nombre de tirages à réaliser :")) #on définit le nombre de tirages à effectuer
      
for i in range(1,nb_tirage+1): #début et paramétrage de la boucle
      result=randint(inf,sup)  #on stocke le résultat de tirage dans la variable "result"
      print(result)            #on affiche le contenu de "result" à chaque tour de boucle
print("fin de boucle")         #dès que i = nb_tirage+1 on sort de la boucle et on l'affiche
```
