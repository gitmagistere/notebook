## Parcourir une liste

### simple

mettons nous dans le cas où nous désirons parcourir une liste de valeurs expérimentales quelconques

```python
# définition de la liste
valeurs = [32,38,5,41.3,49.7, 57.6,66.6,74.9,92.4]

# Affichage des éléments de la liste un à un, vous remarquerez que l'on ne donne aucune valeur
# particulière initiale à l'objet d'itération "element"
for element in valeurs:
    print(element)
```

### un peu plus détaillé

On peut également utiliser une fonction particulière, à savoir **enumerate()** qui donne la position de l'objet affiché dans la liste

```python
# définition de la liste
valeurs = [32,38,5,41.3,49.7, 57.6,66.6,74.9,92.4]

# Affichage des éléments de la liste un à un avec leur indice
for i, element in enumerate(valeurs):
    print(i, element)
```
