## Travailler sur des tirages aléatoires

!!! info tirer au sort

Python possède des bibliothèques permettant de travailler sur le hasard.

!!! abstract tirer au sort: la bibliothèque indispensable

La bibliothèque dédiée en python qui permet de faire de l'aléatoire est **random**
on l'appelle en début de programme comme d'habitude pour les bibliothèques
!!!

```python
# exemple de fonctionnement basique

from random import *

#on tire un nombre aléatoire entier compris entre 1 et 10
tirage=randint(1,10)
print(tirage)
```

on peut également tirer au sort un objet dans une liste d'objets

```python
# exemple de tirage dans une liste
from random import *

liste=['bleu','vert','rouge','cyan','magenta','jaune']
tirage=choice(liste)
print(tirage)
```

!!! abstract ajouter des objets à une liste
Il peut être intéressant d'être capable d'ajouter des objets à une liste vide. Par exemple on peut songer à vouloir stocker le résultat d'un calcul (une composante de vitesse par exemple) afin de pouvoir s'enservir pour afficher une courbe particulière.
!!!

```python
i=0
resultat=[]           #on crée une liste vide
while i<8:            #on initialise la boucle
    y=2*i+1           #on exécute un calcul
    resultat.append(y)#on stocke le résultat par la méthode append dans la liste 
    i=i+1             #on incrémente l'index
print(resultat)       #on affiche le contenu complet de la liste une fois que la boucle est terminée.
```
