# Premiers graphiques

Nous allons nous servir de **Python** pour tracer des graphiques.

Pour ce faire, il est n√©cessaire de demander √† **Python** de charger le module **pyplot** de la librairie **matplotlib**, module d√©di√© aux trac√©s de courbes. 

Les fonctions que nous appellerons seront pr√©c√©d√©es par "**plt.**".

Nous allons, par exemple, tracer la fonction f d√©finie par $f (x) = x^2$ .


```python
import matplotlib.pyplot as plt # le module pyplot est renomm√© plt
import numpy as np # le module numpy est renomm√© np
x=np.linspace(1,10,100) # cr√©e une liste de 100 nombres sur l'intervalle [1; 10]
y=x**2 #fonction x au carr√©

#D√©but de la figure graphique
plt.figure(1) #nom du graphique
plt.clf() #nettoyage de la figure
plt.plot(x,y) # Trace la fonction mais ne montre pas le graphique.
plt.grid(True) #Affichage d'une grille
plt.show
```

Le fait de nommer la figure et de la nettoyer, nous sera utile si nous faisons tracer plusieurs courbes dans le m√™me programme.

L'affichage d'un quadrillage (grille) se fait gr√¢ce √† la commande **plt.grid(True)**.

Si tout se passe correctement, vous devriez voir votre graphe avec un axe horizontal, un axe vertical et la courbe en bleu. S'il y a plusieurs courbes, chacune aura une couleur diff√©rente.

```python
import matplotlib.pyplot as plt # le module pyplot est renomm√© plt
import numpy as np # le module numpy est renomm√© np
x=np.linspace(1,10,100) 
y=x**2 

#D√©but de la figure graphique
plt.figure(1) 
plt.clf() 
plt.plot(x,y) 
plt.grid(True)
plt.savefig('premier_graphique',dpi=350)
plt.show()
```

*Si nous voulons pouvoir √©crire des applications v√©ritablement utiles, il nous faut des techniques permettant d‚Äô aiguiller le d√©roulement du programme dans diff√©rentes directions, en fonction des circonstances rencontr√©es. *L‚Äôinstruction **if**(si en fran√ßais) est la structure de test la plus simple. Sa syntaxe en **Python** fait intervenir la notion de bloc. Nous allons d‚Äôabord √©tudier cette notion plus en d√©tail.*

## Bloc et indentation

Un bloc est d√©fini par une indentation obtenue en d√©calant le d√©but des instructions vers la droite gr√¢ce √† des espaces en d√©but de ligne (4 espaces sous Windows). Toutes les instructions d‚Äôun m√™me bloc doivent √™tre indent√©es exactement au m√™me niveau (c‚Äôest-√†-dire d√©cal√©es √† droite d‚Äôun m√™me nombre d‚Äôespaces).

Un bloc peut contenir une ou plusieurs instructions, et notamment des instructions compos√©es (tests, boucles, etc.).

## Commande If

%% figure

Ici **condition** est une expression bool√©enne, c‚Äôest-√†-dire une expression qui prend pour valeur **True** (Vrai) ou **False** (Faux).

L'instruction **A** n‚Äôest ex√©cut√©e que si la condition est v√©rifi√©e (c‚Äôest-√†-dire si elle prend pour valeur **True**).

L‚Äôinstruction **if** est une instruction compos√©e. Le **:** (deux-points) √† la fin de la ligne introduit le bloc d‚Äôinstructions qui sera ex√©cut√© si la condition est v√©rifi√©e.

>Exemple 1 avec condition vraie:


```python
x = 15
if x > 10:
    print(x, "est plus grand que 10")
print("Fin")
```

>Exemple 1 avec condition fausse :

```python
x = 3
if x > 10:
    print x, "est plus grand que 10"
print("Fin")
```

>Exemple 2 avec condition vraie :

```python
x = 5
if x > 0:
    print(x, "est plus grand que 0")
    print("il est strictement positif")
print("Fin")
```

## Commande If ... Else

L'instruction **else** (sinon) permet de rajouter une instruction si la condition du **if** n'est  pas remplie (**false**).

%%% figure

La syntaxe de cette commande est:

```python
if condition:
    Instruction A
else:
    Instruction B
```

Exemple o√π la condition est vraie :

```python
x = 5
if x > 0:
    print(x, "est positif")
else:
    print(x, "est n√©gatif ou nul")
print("Fin")
```

Exemple o√π la condition est fausse :

```python
x = -2
if x > 0:
    print(x, "est positif")
else:
    print(x, "est n√©gatif ou nul")
print("Fin")
```

Une expression bool√©enne peut contenir les op√©rateurs de comparaison suivants :

Op√©rateur|Signification
---------|:----------:

$x!=y$ |$x$ est diff√©rent de $y$
$x>y$ |$x$ est sup√©rieur √† $y$
$x<y$ | $x$ est inf√©rieur √† $y$
$x>=y$ |$x$ est sup√©rieur ou √©gal √† $y$
$x<=y$ |$x$ est inf√©rieur ou √©gal √† $y$

Attention cependant √† la condition d'√©galit√© qui risque de ne jamais √™tre atteinte avec des nombres flottants (de type float). On pr√©f√©rera comparer les valeurs de deux nombres ou comparer leur diff√©rence √† un nombre tout petit.

# Notion de boucle

*L‚Äôune des t√¢ches que les machines font le mieux est la r√©p√©tition sans erreur de t√¢ches identiques. Il existe bien des m√©thodes pour programmer ces t√¢ches r√©p√©titives. Nous allons commencer par l‚Äôune des plus fondamentales: la boucle de r√©p√©tition construite autour de l‚Äôinstruction **while** puis nous aborderons l'instruction **for**.*

##While

Il est souvent n√©cessaire de r√©p√©ter plusieurs fois une m√™me s√©quence d'instructions jusqu'√† ce qu'une certaine condition soit satisfaite. Pour se faire, on utilise une boucle.
Le mot-cl√© **while** (tant que) permet de r√©aliser cela. Vous trouverez, ci-dessous, un exemple complet de la structure d'une boucle.


```python
instruction 1
while condition A : # se termine par :
    instruction 2 # indentation
    instruction 3 # indentation
instruction 4
```

Dans le programme ci-dessus, l'instruction 1 est ex√©cut√©e puis les instructions 2 et 3 sont r√©p√©t√©es tant que la condition A est vraie (**True**). Quand la condition A devient fausse (**False**), on passe √† l'instruction 4. Si la condition A est fausse d√®s le d√©part, les instructions 2 et 3 ne sont jamais ex√©cut√©es. 

La ligne du bloc qui commence par **while** se termine par ":" et les lignes suivantes sont indent√©es.

Si on veut ex√©cuter les instructions 2 et 3 quand la condition est fausse, on doit ajouter **not** apr√®s while dans l'algorithme ci-dessus et si la condition est vraie alors c'est l'instruction 4 qui sera seulement ex√©cut√©e. 

```python
instruction 1
while True : # se termine par :
    instruction 2 #indentation
    instruction 3 #indentation
instruction 4
```

## For ... in ...

Cette structure de boucle √©vite de devoir d√©finir et incr√©menter un compteur, la variable cr√©√©e juste apr√®s la commande **for** √©tant locale. La grande diff√©rence avec le mot-cl√© **while** r√©side dans le fait que le nombre d'it√©rations est pr√©d√©fini.

La repr√©sentation des vecteurs va surtout nous servir en Physique-Chimie pour tracer des vecteurs vitesse et acc√©l√©ration.

Le script suivant exploite une chronophotographie fournie dans le fichier *ParaboleAvimeca.txt*:

Le vecteur vitesse $\overrightarrow{v}$ sera approch√© √† chaque instant par le vecteur vitesse moyenne:

$\overrightarrow{v}_i = \dfrac{\overrightarrow{OM_{i+1}} - \overrightarrow{OM_{i}}}{\Delta t} = \dfrac{\overrightarrow{M_{i}M_{i+1}}}{\Delta t}$

et l'acc√©l√©ration $\overrightarrow{a}$ par:

$\overrightarrow{a}_i = \dfrac{\overrightarrow{v_{i+1}} - \overrightarrow{v_{i}}}{\Delta t}$


```python
import csv #ouverture du package csv
import matplotlib.pyplot as plt

'''D√©finition de la fonction LectureFichier permettant d'extraire les donn√©es issues d'Avimeca'''

def LectureFichier(fichier,separateur,n):
    file = open(fichier,"r") #ouverture du fichier en mode lecture
    reader = csv.reader(file,delimiter=separateur)
    colonne = [] #cr√©ation d'une liste colonne qui contiendra les donn√©es d'une colonne du fichier CSV
    for row in reader:
        try: #assertion qui permet de mettre les valeurs num√©riques au bon format
            notation_point = row[n].replace(",",".") #remplace la virgule par un point
            colonne.append(float(notation_point)) #ajoute la donn√©e lue dans la liste colonne en tant que nombre et non plus comme cha√Æne de caract√®re
        except: # Rien ne doit se passer en cas d'erreur
            pass
    file.close()
    return colonne

''' Extraction des dates, abscisses et ordonn√©es du fichier g√©n√©r√© par avim√©ca'''

temps = LectureFichier("ParaboleAvimeca.txt",";",0) # 0 correspond √† la premi√®re colonne
abscisses = LectureFichier("ParaboleAvimeca.txt",";",1) # 1 correspond √† la colonne 2
ordonnees = LectureFichier("ParaboleAvimeca.txt",";",2) # 1 correspond √† la colonne 3

Echelle_Variation_Vitesse = 1.0
Echelle_Vitesse = 0.2

#Nombre de points contenus dans l'acquisition AVIMECA
N_points=len(temps)
#Intervalle de temps entre chaque image
delta_t = temps[1]-temps[0]
#print(delta_t)

#Calcul de la vitesse horizontale entre le premier et l'avant dernier point
vx = ['D']
for i in range(0,N_points-1):
    vx.append((abscisses[i+1]-abscisses[i])/(delta_t)
vx.append('F')

#Calcul de la vitesse verticale entre le premier et l'avant dernier point
vy = ['D']
for i in range(0,N_points-1):
    vy.append((ordonnees[i+1]-ordonnees[i])/delta_t)
vy.append('F')


#Calcul de la variation de vitesse horizontale entre le deuxi√®me et l'avant avant dernier point
delta_vx = ['D']
for i in range(1,N_points-2):
    delta_vx.append(vx[i+1]-vx[i])
delta_vx.append('F')

#Calcul de la variation de vitesse verticale entre le deuxi√®me et l'avant avant dernier point
delta_vy = ['D']
for i in range(1,N_points-2):
    delta_vy.append(vy[i+1]-vy[i])
delta_vy.append('F')


#Trac√© de la trajectoire
#Cr√©ation d'une fen√™tre graphique
plt.figure(1)
#Nettoyage de la fen√™tre
plt.clf()
#Param√©trages du graphique - optimisation des √©chelles
plt.grid()
plt.xlim(0,1.5 * abscisses[N_points-1])
plt.ylim(0, 1.5 * max(ordonnees))
#Cr√©ation du graphique
#plt.plot(abscisses,ordonnees,'r.')
for i in range(1,N_points,4):
    plt.arrow(abscisses[i],ordonnees[i],Echelle_Vitesse * vx[i],Echelle_Vitesse * vy[i],color='r')
#for i in range(2,N_points-1,4):
#    plt.arrow(abscisses[i],ordonnees[i],Echelle_Variation_Vitesse * delta_vx[i],Echelle_Vitesse * delta_vy[i],color='b')
#Affichage du graphique
plt.savefig('1')
plt.show()



#Trac√© de la composante horizontale de la vitesse
#Cr√©ation d'une fen√™tre graphique
plt.figure(2)
#Nettoyage de la fen√™tre
plt.clf()
#Param√©trages du graphique - optimisation des √©chelles
plt.grid()
#plt.xlim(0,1.5 * temps[N_points-1])
plt.ylim(0, 1.1*max(vx[1:N_points-1]))
#Cr√©ation du graphique
plt.plot(temps[1:N_points-1],vx[1:N_points-1],'r.')
plt.savefig('2')
plt.show()


#Trac√© de la composante verticale de la vitesse
#Cr√©ation d'une fen√™tre graphique
plt.figure(3)
#Nettoyage de la fen√™tre
plt.clf()
#Param√©trages du graphique - optimisation des √©chelles
plt.grid()
plt.xlim(0,temps[N_points-1])
plt.ylim(1.1*min(vy[1:N_points-1]), 1.1*max(vy[1:N_points-1]))
#Cr√©ation du graphique
plt.plot(temps[1:N_points-1],vy[1:N_points-1],'r.')
plt.savefig('3')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#cr√©ation de deux variables X et Y
X = np.linspace(-10, 10, 21)
Y = np.linspace(-10, 10, 21)

#Attribution pour chaque point de l'espace d'une 
# abscisse U et d une ordonn√©e V pour le champ de vecteur
U, V = np.meshgrid(X, Y)

#nom de la figure
plt.figure(1)
plt.clf()
fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = 1
longueurOnde = 300
#Attention, le mot  lambda est d√©j√† utilis√© par Python donc on ne peut pas 
# appeler une variable lambda mais on peut utiliser le symbole 
Phi = 3.14
v = 40
T = longueurOnde/v
```

Nous allons faire avancer le temps gr√¢ce √† une boucle **for** et √† chaque temps, nous allons calculer les ordonn√©es de la courbe pour toutes les abscisses allant de 0 √† $x_{\text{max}}= v \times t$. Autrement dit pour chaque temps, on doit calculer en priorit√© jusqu‚Äôo√π l‚Äôonde s‚Äôest propag√©e.

```python
for t in range(200): #le temps va √©voluer de 0 √† 199
	plt.clf()
	x_max = v*t #on calcule l'abscisse de la propagation
	y = [] #on initialise la liste des ordonn√©es
	x = [] #pareil pour les abscisses
	for xi in range(0, x_max):
		x.append(xi) # on ajoute xi √† la liste des abscisses
		y.append(A*np.sin((2*3.14/longueurOnde)*xi-(2*3.14/T)*t+Phi)) # dans la ligne pr√©c√©dente, on calcule l'ordonn√©e en xi
	x.append(1500) # on rajoute le point 1500,0 pour afficher un trait o√π l'onde ne s'est pas encore propag√©e
	y.append(0)
	plt.axis([-1, 1500, -2, 2])
	plt.plot(x, y, 'r-')
	plt.pause(0.01)
plt.show()
```

## R√©solutions d'√©quations avec le module \Pyth{scipy}

Le module **scipy**, comme **scientifique Python**, va permettre assez facilement de r√©soudre des √©quations via la fonction **fsolve**.

Le code suivant permet de trouver les solutions de l'√©quation math√©matique suivante: $f(x)=x^2-3x-4=0$. 

```python
import scipy.optimize as opt
from numpy import around

f=lambda x: x**2-3*x-4

#r√©solution
sol=opt.fsolve(f,[-6,6]) #f(0)=-4, il y a donc une solution positive, une n√©gative. Les solutions sont recherch√©es sur l'intervalle [-6,6]
print(sol)
print("la solution positive est : ",sol[1])
print("la solution n√©gative est : ",sol[0])
```

Autre possibilit√© : la fonction **newton** ne renvoie qu'une racine 

```python
sol1=opt.newton(f,6)
sol2=opt.newton(f,-6)
```

#Liaison s√©rie

La communication avec le port s√©rie va jouer un r√¥le majeur dans l'acquisition de donn√©es. Lorsque nous travaillerons avec les micro-contr√¥leurs Arduino qui ne sont pas capables de stocker un nombre important de donn√©es, il nous sera utile de pouvoir les r√©cup√©rer pour pouvoir les enregistrer, par exemple dans un fichier **.csv**.

##Le port s√©rie


En communication s√©rie, l'information √† transmettre est d√©coup√©e en petits blocs de taille fixe avant de la transmettre. La taille des blocs correspond au nombre des lignes disponibles pour la transmission des donn√©es.

Ce type de communication s'oppose √† la **communication parall√®le**. En communication parall√®le, il y a une ligne par bits √† transmettre. Tous les bits sont donc transmis en m√™me temps. 

Pour une m√™me fr√©quence de communication, la communication parall√®le est donc plus rapide.

L'avantage de la communication s√©rie sur la communication parall√®le est qu'elle n√©cessite moins de lignes, donc moins de broches, donc moins de composants. Son co√ªt est donc plus faible.

Avec Python 3, les communications doivent se faire avec des objets de type **bytes** (cha√Ænes d'octets). Ceci est une diff√©rence avec Python 2 qui communique avec des objets de type cha√Æne de caract√®res (type **str**).

## R√©cup√©rer les donn√©es du port s√©rie

Le module **serial** est n√©cessaire pour r√©cup√©rer les donn√©es du port s√©rie.

**Remarque:** L‚Äôouverture du port s√©rie peut ne pas r√©ussir pour diverses raisons (mauvais port, etc..) et si l‚Äôon ne veut pas que le programme s‚Äôarr√™te, on peut utiliser une structure sp√©ciale, la structure **try : .. except : ..**. Ce qui permet de g√©rer l‚Äôerreur sans interruption du programme ce qui donne :

```python
import serial
import time

#Connexion Linux au port s√©rie
serial_port=serial.Serial(port="/dev/ttyACM1",baudrate=9600)
serial_port.setDTR(False)
time.sleep(0.1)
serial_port.setDTR(True)
serial_port.flushInput( )
# les mesures
mesure = [ ]
temps = [ ]
serial_port.flushInput( )
for i in range(100) :
    val = serial_port.readline( ).split( ) #permet de couper
    try :
        t = float(val[0])/1000
        m = float(val[1])
        temps.append(t)
        mesure.append(m)
        #print(val)
        print(i,'/100')
    except :
        pass
# fermeture du port s√©rie
serial_port.close()
```

Attention le port s√©lectionn√© change de d√©nomination suivant votre syst√®me d'exploitation:

- Sous Linux : /dev/ttyACM suivi d‚Äôun num√©ro (0 ou 1 en g√©n√©ral) ou /dev/ttyUSB0 ,
- Sous Mac: /dev/cu.usbmodem1432301 ,
- Sous Windows : COM suivi d‚Äôun num√©ro (1, 2, 3,\ldots{}).

**Remarque:** Depuis le terminal, la commande ls /dev peut vous permettre de conna√Ætre la liste des ports utilis√©s par l'ordinateur.

## Sauvegarder les donn√©es dans un fichier csv

```python
import csv

def writeCSV(fichier,sep,col1,nomcol1,col2,nomcol2):
#fichier <str> : Le nom du fichier CSV √† cr√©√©r -> "mon_fichier.csv"
#sep <str> : Le s√©parateur des colonnes
#col1 <list> : La premi√®re colonne -> [1, 1.5, 2, ...]
#nomcol1 : nom de la premi√®re liste
#col2 <list> : La deuxi√®me colonne -> [100, 50, 100, ...]
#nomcol2 : nom de la deuxi√®me liste
	file = open(fichier,"w")
	writer = csv.writer(file,delimiter =sep )
	writer.writerow((nomcol1,nomcol2))
	fin1 , fin2 = len(col1), len(col2)
    if fin1 == fin2 :
        for i in range (fin1) :
            writer.writerow((col1[i],col2[i]))
        else :
            print ( "Les deux listes n ont pas la m√™me taille" )
        file.close( )
        
writeCSV('mesures.csv',',',temps,'temps(ms)',mesure,'tension(V)')
```
