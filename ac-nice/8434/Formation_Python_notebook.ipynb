## Les listes

Une liste est une collection ordonnée et mutable (modifiable) d'objets (type **list**).
- **Collection ordonnée:** le premier élément de la liste est indexé par 0, le second par 1, et ainsi de suite. 
- **Collection mutable:** il est possible, sans réaffectation complète, de changer le contenu de la liste : plus exactement, il est possible d'ajouter un élément, de supprimer un élément et d'étendre la liste par une autre liste. 

Les listes permettent de construire des collections d'objets de proche en proche: ce n'est que par la pratique que vous commencerez à imaginer l'étendue des possibilités! 

### Créer une liste

Il existe plusieurs façons de créer une liste:

```python
L=[1,3,7.4,5]
```

Une liste vide se crée donc ainsi:

```python
L=[]
```

```python
L=list(range(0,101,5))
```

ou

3. Par ajout élément par élément des objets dans une liste initialement vide.
4. En utilisant une liste en compréhension, vue précédemment au chapitre sur les boucles.

# Comparaison de valeurs et test

*Si nous voulons pouvoir écrire des applications véritablement utiles, il nous faut des techniques permettant d’ aiguiller le déroulement du programme dans différentes directions, en fonction des circonstances rencontrées. Pour ce faire, nous devons disposer d’ instructions capables de tester une certaine condition et de modifier le comportement du programme en conséquence.*

*L’instruction **if**(si en français) est la structure de test la plus simple. Sa syntaxe en **Python** fait intervenir la notion de bloc. Nous allons d’abord étudier cette notion plus en détail.*

## Bloc et indentation

Un bloc est défini par une indentation obtenue en décalant le début des instructions vers la droite grâce à des espaces en début de ligne (4 espaces sous Windows). Toutes les instructions d’un même bloc doivent être indentées exactement au même niveau (c’est-à-dire décalées à droite d’un même nombre d’espaces).

Un bloc peut contenir une ou plusieurs instructions, et notamment des instructions composées (tests, boucles, etc.).

## Commande If

%% figure

Ici **condition** est une expression booléenne, c’est-à-dire une expression qui prend pour valeur **True** (Vrai) ou **False** (Faux).

L'instruction **A** n’est exécutée que si la condition est vérifiée (c’est-à-dire si elle prend pour valeur **True**).

L’instruction **if** est une instruction composée. Le **:** (deux-points) à la fin de la ligne introduit le bloc d’instructions qui sera exécuté si la condition est vérifiée.

>Exemple 1 avec condition vraie:


```python
x = 15
if x > 10:
    print(x, "est plus grand que 10")
print("Fin")
```

>Exemple 1 avec condition fausse :

```python
x = 3
if x > 10:
    print x, "est plus grand que 10"
print("Fin")
```

>Exemple 2 avec condition vraie :

```python
x = 5
if x > 0:
    print(x, "est plus grand que 0")
    print("il est strictement positif")
print("Fin")
```

## Commande If ... Else

L'instruction **else** (sinon) permet de rajouter une instruction si la condition du **if** n'est  pas remplie (**false**).

%%% figure

La syntaxe de cette commande est:

```python
if condition:
    Instruction A
else:
    Instruction B
```

Exemple où la condition est vraie :

```python
x = 5
if x > 0:
    print(x, "est positif")
else:
    print(x, "est négatif ou nul")
print("Fin")
```

Exemple où la condition est fausse :

```python
x = -2
if x > 0:
    print(x, "est positif")
else:
    print(x, "est négatif ou nul")
print("Fin")
```

Une expression booléenne peut contenir les opérateurs de comparaison suivants :

Opérateur|Signification
---------|:----------:
$x==y$| $x$ est égal à $y$
$x!=y$ |$x$ est différent de $y$
$x>y$ |$x$ est supérieur à $y$
$x<y$ | $x$ est inférieur à $y$
$x>=y$ |$x$ est supérieur ou égal à $y$
$x<=y$ |$x$ est inférieur ou égal à $y$

Attention cependant à la condition d'égalité qui risque de ne jamais être atteinte avec des nombres flottants (de type float). On préférera comparer les valeurs de deux nombres ou comparer leur différence à un nombre tout petit.

# Notion de boucle

*L’une des tâches que les machines font le mieux est la répétition sans erreur de tâches identiques. Il existe bien des méthodes pour programmer ces tâches répétitives. Nous allons commencer par l’une des plus fondamentales: la boucle de répétition construite autour de l’instruction **while** puis nous aborderons l'instruction **for**.*

##While

Il est souvent nécessaire de répéter plusieurs fois une même séquence d'instructions jusqu'à ce qu'une certaine condition soit satisfaite. Pour se faire, on utilise une boucle.
Le mot-clé **while** (tant que) permet de réaliser cela. Vous trouverez, ci-dessous, un exemple complet de la structure d'une boucle.


```python
instruction 1
while condition A : # se termine par :
    instruction 2 # indentation
    instruction 3 # indentation
instruction 4
```

Dans le programme ci-dessus, l'instruction 1 est exécutée puis les instructions 2 et 3 sont répétées tant que la condition A est vraie (**True**). Quand la condition A devient fausse (**False**), on passe à l'instruction 4. Si la condition A est fausse dès le départ, les instructions 2 et 3 ne sont jamais exécutées. 

La ligne du bloc qui commence par **while** se termine par ":" et les lignes suivantes sont indentées.

Si on veut exécuter les instructions 2 et 3 quand la condition est fausse, on doit ajouter **not** après while dans l'algorithme ci-dessus et si la condition est vraie alors c'est l'instruction 4 qui sera seulement exécutée. 

```python
instruction 1
while True : # se termine par :
    instruction 2 #indentation
    instruction 3 #indentation
instruction 4
```

## For ... in ...

Une autre façon de faire une boucle est d'utiliser les mot-clés **for** et **in** quand on doit parcourir une séquence (chaîne de caractères ou liste). Cette structure de boucle évite de devoir définir et incrémenter un compteur, la variable créée juste après la commande **for** étant locale. La grande différence avec le mot-clé **while** réside dans le fait que le nombre d'itérations est prédéfini.

Voici des exemples pour mieux comprendre la syntaxe. Les possibilités sont nombreuses. 

> *Exemple 1:* La table de 9

```python
for i in range(11) :
    print(i," *9= ",9*i)
```

Ici, la fonction **range** crée un itérateur (ne stockant qu'un seul nombre en mémoire) dont les valeurs entière varient entre 0 et 10 inclus.

> *Exemple 2:* Carré des 100 premiers nombres impairs


```python
for i in range(100) :
    print((2*i+1)**2)
```

## Les listes en compréhension

C'est la méthode **la plus rapide** pour créer une liste : plus que les méthodes (par exemple avec append), qui étaient déjà plus rapides qu'une concaténation d'autres listes. Au final, construire une liste en compréhension permet de remplacer une ou plusieurs boucles **for** et l'évaluation d'une condition **if** par une seule ligne de code : celui-ci est rendu particulièrement compacte mais parfois peu lisible, ce qui n'est pas un avantage.

Les exemples sont parlants:

>**Exemple 1**:

```python
liste=[]
for i in range(10)
    liste.append(i)
```

est équivalent à:

```python
liste=[i for i in range(10)]
```

En sortie, **Python** renvoie la liste des entiers de 0 (inclus) à 10 (exclu).

>**Exemple 2**:

```python
liste=[]
for i in range(20)
    if i%2==1:
        liste.append(i)
```

est équivalent à:

```python
liste=[i for i in range(20) if i%2==1] # i modulo 2=1
```

En sortie, **Python** renvoie la liste des entiers impairs de 0 (inclus) à 20 (exclu).

>**Exemple 3**:

```python
liste=[]
for i in range(20)
    liste.append((i+1)**2)
```

est équivalent à:

```python
liste=[(i+1)**2 for i in range(20)
```

En sortie, **Python** renvoie la liste des carrés des entiers de 1 (inclus) à 20 (inclus).

Lorsque l' on écrit des programmes, nous sommes souvent amenés à répéter plusieurs fois une même séquence d' instructions. C' est fastidieux, surtout que des erreurs ne manquent pas de se produire. 
Pour cela, le langage dispose d' une syntaxe qui permet de regrouper une séquence et de la désigner par un mot unique, une **fonction**. 
Une fonction est une portion de code réutilisable que l' on va regrouper pour lui donner un nom. 
On pourra exécuter de nouveau cette portion en appelant simplement le nom de la fonction et ses arguments.

## Syntaxe de déclaration de fonction




Voici la syntaxe de déclaration d' une fonction

```python
def nom_de_Fonction (parametre1, parametre2, ...) :
    #documentation de la fonction
    instruction 1
    instruction 2
    ...
    return (parametreDeSortie1, parametreDeSortie2, ...)
```

La documentation est facultative mais vraiment conseillée, **return** peut aussi être absent, ce que l' on appelle parfois une procédure.

Après avoir exécuter l' instruction **return**, le programme sort du bloc de la fonction : aucune autre instruction de la fonction ne sera exécutée après **return**.

**Remarque fondamentale**:**return** renvoie un objet (entier, flottant, ...,ou une collection comme un tuple, une liste), mais ne l' affiche pas (c' est le rôle de **print**). En revanche, un autre programme peut utiliser le contenu renvoyé par **return**, ce qui n' est pas possible avec **print**.

 <font color='red'>Ne confondez pas les deux termes et surtout leurs utilisations!</font>

Voici maintenant un exemple de procédure permettant d' afficher une table de multiplication, avec comme arguments un nombre de départ, un de fin et la base.

```python
def table(base,debut,fin):
    # Affiche la table de multiplication de <base> de <début> à <fin>
    n=debut
    while n<=fin:
        print(n,' *' ,base,' =' ,n*base)
        n+=1
```

##Fichiers csv ou txt

Le principe de l'importation consiste à utiliser une fonction (**LectureFichier**) qui va permettre de stocker les valeurs numériques dans des listes (ici **colonne**).
Le fichier **txt** ou **csv** est parcouru, et à chaque nouvelle ligne du fichier, une nouvelle valeur est rajoutée dans la liste créée via la commande **append**. Une conversion en nombre flottant est alors nécessaire.


```python
import csv #importation du module csv


'''Définition de la fonction LectureFichier permettant d'extraire les données issues d'Avimeca'''

def LectureFichier(fichier,separateur,n):
    file = open(fichier,"r") #ouverture du fichier en mode lecture
    reader = csv.reader(file,delimiter=separateur)
    colonne = [] #création d'une liste colonne qui contiendra les données d'une colonne du fichier CSV
    for row in reader:
        try: #assertion qui permet de mettre les valeurs numériques au bon format
            notation_point = row[n].replace(",",".") #remplace la virgule par un point
            colonne.append(float(notation_point)) #ajoute la donnée lue dans la liste colonne en tant que nombre et non plus comme chaîne de caractère
        except: # Rien ne doit se passer en cas d'erreur
            pass
    file.close()
    return colonne

''' Extraction des dates, abscisses et ordonnées du fichier généré par aviméca'''

temps = LectureFichier("ParaboleAvimeca.txt",";",0) # 0 correspond à la première colonne
abscisses = LectureFichier("ParaboleAvimeca.txt",";",1) # 1 correspond à la colonne 2
ordonnees = LectureFichier("ParaboleAvimeca.txt",";",2) # 1 correspond à la colonne 3
```

La représentation des vecteurs va surtout nous servir en Physique-Chimie pour tracer des vecteurs vitesse et accélération.

Le script suivant exploite une chronophotographie fournie dans le fichier *ParaboleAvimeca.txt*:

Le vecteur vitesse $\overrightarrow{v}$ sera approché à chaque instant par le vecteur vitesse moyenne:

$\overrightarrow{v}_i = \dfrac{\overrightarrow{OM_{i+1}} - \overrightarrow{OM_{i}}}{\Delta t} = \dfrac{\overrightarrow{M_{i}M_{i+1}}}{\Delta t}$

et l'accélération $\overrightarrow{a}$ par:

$\overrightarrow{a}_i = \dfrac{\overrightarrow{v_{i+1}} - \overrightarrow{v_{i}}}{\Delta t}$


```python
import csv #ouverture du package csv
import matplotlib.pyplot as plt

'''Définition de la fonction LectureFichier permettant d'extraire les données issues d'Avimeca'''

def LectureFichier(fichier,separateur,n):
    file = open(fichier,"r") #ouverture du fichier en mode lecture
    reader = csv.reader(file,delimiter=separateur)
    colonne = [] #création d'une liste colonne qui contiendra les données d'une colonne du fichier CSV
    for row in reader:
        try: #assertion qui permet de mettre les valeurs numériques au bon format
            notation_point = row[n].replace(",",".") #remplace la virgule par un point
            colonne.append(float(notation_point)) #ajoute la donnée lue dans la liste colonne en tant que nombre et non plus comme chaîne de caractère
        except: # Rien ne doit se passer en cas d'erreur
            pass
    file.close()
    return colonne

''' Extraction des dates, abscisses et ordonnées du fichier généré par aviméca'''

temps = LectureFichier("ParaboleAvimeca.txt",";",0) # 0 correspond à la première colonne
abscisses = LectureFichier("ParaboleAvimeca.txt",";",1) # 1 correspond à la colonne 2
ordonnees = LectureFichier("ParaboleAvimeca.txt",";",2) # 1 correspond à la colonne 3

Echelle_Variation_Vitesse = 1.0
Echelle_Vitesse = 0.2

#Nombre de points contenus dans l'acquisition AVIMECA
N_points=len(temps)
#Intervalle de temps entre chaque image
delta_t = temps[1]-temps[0]
#print(delta_t)

#Calcul de la vitesse horizontale entre le premier et l'avant dernier point
vx = ['D']
for i in range(0,N_points-1):
    vx.append((abscisses[i+1]-abscisses[i])/(delta_t)
vx.append('F')

#Calcul de la vitesse verticale entre le premier et l'avant dernier point
vy = ['D']
for i in range(0,N_points-1):
    vy.append((ordonnees[i+1]-ordonnees[i])/delta_t)
vy.append('F')


#Calcul de la variation de vitesse horizontale entre le deuxième et l'avant avant dernier point
delta_vx = ['D']
for i in range(1,N_points-2):
    delta_vx.append(vx[i+1]-vx[i])
delta_vx.append('F')

#Calcul de la variation de vitesse verticale entre le deuxième et l'avant avant dernier point
delta_vy = ['D']
for i in range(1,N_points-2):
    delta_vy.append(vy[i+1]-vy[i])
delta_vy.append('F')


#Tracé de la trajectoire
#Création d'une fenêtre graphique
plt.figure(1)
#Nettoyage de la fenêtre
plt.clf()
#Paramétrages du graphique - optimisation des échelles
plt.grid()
plt.xlim(0,1.5 * abscisses[N_points-1])
plt.ylim(0, 1.5 * max(ordonnees))
#Création du graphique
#plt.plot(abscisses,ordonnees,'r.')
for i in range(1,N_points,4):
    plt.arrow(abscisses[i],ordonnees[i],Echelle_Vitesse * vx[i],Echelle_Vitesse * vy[i],color='r')
#for i in range(2,N_points-1,4):
#    plt.arrow(abscisses[i],ordonnees[i],Echelle_Variation_Vitesse * delta_vx[i],Echelle_Vitesse * delta_vy[i],color='b')
#Affichage du graphique
plt.savefig('1')
plt.show()



#Tracé de la composante horizontale de la vitesse
#Création d'une fenêtre graphique
plt.figure(2)
#Nettoyage de la fenêtre
plt.clf()
#Paramétrages du graphique - optimisation des échelles
plt.grid()
#plt.xlim(0,1.5 * temps[N_points-1])
plt.ylim(0, 1.1*max(vx[1:N_points-1]))
#Création du graphique
plt.plot(temps[1:N_points-1],vx[1:N_points-1],'r.')
plt.savefig('2')
plt.show()


#Tracé de la composante verticale de la vitesse
#Création d'une fenêtre graphique
plt.figure(3)
#Nettoyage de la fenêtre
plt.clf()
#Paramétrages du graphique - optimisation des échelles
plt.grid()
plt.xlim(0,temps[N_points-1])
plt.ylim(1.1*min(vy[1:N_points-1]), 1.1*max(vy[1:N_points-1]))
#Création du graphique
plt.plot(temps[1:N_points-1],vy[1:N_points-1],'r.')
plt.savefig('3')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = 1
longueurOnde = 300
#Attention, le mot  lambda est déjà utilisé par Python donc on ne peut pas 
# appeler une variable lambda mais on peut utiliser le symbole 
Phi = 3.14
v = 40
T = longueurOnde/v
```

Nous allons faire avancer le temps grâce à une boucle **for** et à chaque temps, nous allons calculer les ordonnées de la courbe pour toutes les abscisses allant de 0 à $x_{\text{max}}= v \times t$. Autrement dit pour chaque temps, on doit calculer en priorité jusqu’où l’onde s’est propagée.

```python
for t in range(200): #le temps va évoluer de 0 à 199
	plt.clf()
	x_max = v*t #on calcule l'abscisse de la propagation
	y = [] #on initialise la liste des ordonnées
	x = [] #pareil pour les abscisses
	for xi in range(0, x_max):
		x.append(xi) # on ajoute xi à la liste des abscisses
		y.append(A*np.sin((2*3.14/longueurOnde)*xi-(2*3.14/T)*t+Phi)) # dans la ligne précédente, on calcule l'ordonnée en xi
	x.append(1500) # on rajoute le point 1500,0 pour afficher un trait où l'onde ne s'est pas encore propagée
	y.append(0)
	plt.axis([-1, 1500, -2, 2])
	plt.plot(x, y, 'r-')
	plt.pause(0.01)
plt.show()
```

```python
import scipy.optimize as opt
from numpy import around

f=lambda x: x**2-3*x-4

#résolution
sol=opt.fsolve(f,[-6,6]) #f(0)=-4, il y a donc une solution positive, une négative. Les solutions sont recherchées sur l'intervalle [-6,6]
print(sol)
print("la solution positive est : ",sol[1])
print("la solution négative est : ",sol[0])
```

#Liaison série

La communication avec le port série va jouer un rôle majeur dans l'acquisition de données. Lorsque nous travaillerons avec les micro-contrôleurs Arduino qui ne sont pas capables de stocker un nombre important de données, il nous sera utile de pouvoir les récupérer pour pouvoir les enregistrer, par exemple dans un fichier **.csv**.

##Le port série


En communication série, l'information à transmettre est découpée en petits blocs de taille fixe avant de la transmettre. La taille des blocs correspond au nombre des lignes disponibles pour la transmission des données.

Ce type de communication s'oppose à la **communication parallèle**. En communication parallèle, il y a une ligne par bits à transmettre. Tous les bits sont donc transmis en même temps. 

Pour une même fréquence de communication, la communication parallèle est donc plus rapide.

L'avantage de la communication série sur la communication parallèle est qu'elle nécessite moins de lignes, donc moins de broches, donc moins de composants. Son coût est donc plus faible.

Avec Python 3, les communications doivent se faire avec des objets de type **bytes** (chaînes d'octets). Ceci est une différence avec Python 2 qui communique avec des objets de type chaîne de caractères (type **str**).

## Récupérer les données du port série

Le module **serial** est nécessaire pour récupérer les données du port série.

**Remarque:** L’ouverture du port série peut ne pas réussir pour diverses raisons (mauvais port, etc..) et si l’on ne veut pas que le programme s’arrête, on peut utiliser une structure spéciale, la structure **try : .. except : ..**. Ce qui permet de gérer l’erreur sans interruption du programme ce qui donne :

```python
import serial
import time

#Connexion Linux au port série
serial_port=serial.Serial(port="/dev/ttyACM1",baudrate=9600)
serial_port.setDTR(False)
time.sleep(0.1)
serial_port.setDTR(True)
serial_port.flushInput( )
# les mesures
mesure = [ ]
temps = [ ]
serial_port.flushInput( )
for i in range(100) :
    val = serial_port.readline( ).split( ) #permet de couper
    try :
        t = float(val[0])/1000
        m = float(val[1])
        temps.append(t)
        mesure.append(m)
        #print(val)
        print(i,'/100')
    except :
        pass
# fermeture du port série
serial_port.close()
```

Attention le port sélectionné change de dénomination suivant votre système d'exploitation:

- Sous Linux : /dev/ttyACM suivi d’un numéro (0 ou 1 en général) ou /dev/ttyUSB0 ,
- Sous Mac: /dev/cu.usbmodem1432301 ,
- Sous Windows : COM suivi d’un numéro (1, 2, 3,\ldots{}).

**Remarque:** Depuis le terminal, la commande ls /dev peut vous permettre de connaître la liste des ports utilisés par l'ordinateur.
