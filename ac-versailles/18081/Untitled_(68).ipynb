<h2> <span style="background:#99CC66;color:#ffffff;"> Édition et exécution d'une cellule de code</span> </h2>

Un notebook jupyter consiste en **cellules** individuelles pouvant contenir du **code** Python, du **texte** (avec formatage Markdown, HTML et LaTeX) ou des **graphiques**. Commençons par les cellules de type Code.

**1)** Placez le curseur dans la 1ère cellule **In[ ]:** ci-dessous puis frappez `<enter>`. Que se passe-t-il ? Rien d'autre que l'insertion d'un saut de ligne... C'est normal, car dans une cellule de type "Code" (c'est-à-dire préfixée par **In[ ]:**) on est par défaut en mode **édition de code** Python !


Pour **exécuter le code** Python d'une cellule, il faut en effet frapper `<maj-entrée>` (ou cliquer sur le bouton <button class='fa fa-step-forward icon-step-forward'></button>, ou encore utiliser le menu `Cellule > Exécuter les cellules et sélectionner la suivante`). Essayez !

```python
15 + 24
```

```python
'Hello tout le monde !'
```

Une ligne **Out[ ]:** devrait être apparue en dessous de la 1ère expression, affichant le résultat de celle-ci ! Le curseur est alors automatiquement passé dans la cellule d'en-dessous que vous pouvez exécuter à son tour.

_Notez qu'il est aussi possible de frapper `<ctrl-entrée>` (identique à `Cellule > Exécuter les cellules `). Dans ce cas, le curseur **reste dans la cellule** après l'exécution du code._

**2)** Exécutez maintenant le code de la cellule ci-dessous.

```python
a = 6     # affectation de la valeur 6 à la variable a
b = 10    # affectation de la valeur 10 à la variable b

print('La somme de ces 2 nombres vaut :', a + b)
```

Remarquez que l'exécution du code ci-dessus n'a retourné aucune valeur. Il n'y a pas de prompt Out[ ]. 

**3)** Peut-on **modifier le code** d'une cellule puis le **ré-exécuter** ? Bien sûr !  
Expérimentez cela dans la cellule ci-dessus en modifiant la valeur des variables a et b, puis frappez à nouveau `<maj-entrée>`

<h2> <span style="background:#99CC66;color:#ffffff;"> Création de cellules de code</span> </h2>

**4)** Comment puis-je créer **mes propres cellules** de code ? Tout simplement de la façon suivante :

- sélectionnez d'abord la cellule juste au-dessus de l'emplacement où vous désirez insérer votre propre cellule  
_(ou respectivement juste en-dessous)_
- puis cliquez sur le bouton <button class='fa fa-plus icon-plus btn btn-xs btn-default'></button> ou faites `Insérer>Insérer une cellule après`  
_(ou respectivement `Insérer>Insérer une cellule avant`)_

Exercez-vous en créant, juste en-dessous de la présente cellule, une nouvelle cellule. 

Insérez-y le code :

    print('Ouaouh...ma première cellule !')

puis exécutez le code de cette cellule.

<h2> <span style="background:#99CC66;color:#ffffff;"> Création de cellules de texte</span> </h2>

Il existe 2 types de cellules de texte, selon le choix effectué à l'aide du menu déroulant de droite ou avec `Cellule > Type de cellule >...` :
- **Texte Brut** : cellules de texte sans aucun formatage, directement éditables sans devoir _double-cliquer_ dedans ;
- **Markdown** : cellules dans lesquelles le texte est soumis à un formatage selon la syntaxe Markdown ; les éventuelles balises HTML et le code LaTeX sont également interprétés...


**7)** Créez juste ci-dessous une cellule de type **Markdown** . Pour cela, créez une cellule par le procédé standard (elle sera donc de type Code), puis changez son type à l'aide du menu déroulant.
