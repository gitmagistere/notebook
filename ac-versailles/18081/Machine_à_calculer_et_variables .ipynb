<h2 style="color:rgb(255,0,0)">Bilan sur les variables</h2>
<div class="alert alert-danger" style="border-left: 15px solid #a94442;border-radius:2px;">
Une variable est un espace mémoire où le programme stocke une valeur. Une variable permet de mémoriser une information.
Les variables possèdent un nom qui est associé à une valeur.  La valeur peut être de nature différente : nombre, texte, etc. On dit que les valeurs des variables sont typées.

Une variable contient une valeur qui peut varier au cours de l’exécution du programme.
Une variable informatique peut être vue comme une boite avec un nom  (étiquette) qui peut contenir différentes valeurs.


Une instruction d’affectation comme  i = 12 se lit i prend pour valeur 12 ou i reçoit la valeur 12. Le nombre 12 est mis en mémoire. Ensuite on  associe la variable i à 12.

Une instruction d’affectation comme y =  x + 4 est exécutée de la façon suivante :
1. Evaluer l’expression x + 4 en ajoutant 4 à la valeur contenue dans la variable x, le résultat est mis en mémoire.
2. Mettre le résultat dans la variable y. Le résultat est associé à la variable y.


**Dans une affectation, le membre de gauche reçoit la valeur du membre de droite.** 
Attention : donner des noms de variables explicites, sans espace.
</div>
