<h2 style="color: orange; font-style: 16px italic verdana;"> Fusionner les tables</h2>

<div class="alert alert-success" style="border-left: 5px solid #3c763d;border-radius:2px;">Ecrire le code pour ouvrir la 2eme table nommée "users_tv.csv" et créer le dataframe <code>users</code>.</div>

```python
users= pd.read_csv("users_tv.csv",sep=";",encoding='utf8')
users
```

La fonction `merge` est une jointure (vous verrez cela en cours de Base de Données à l'automne prochain). Elle permet de créer un nouveau dataframe qui en mélange deux.

le nouveau dataframe a l'ensemble des colonnes des deux dataframe.
   pour les lignes, il combine les lignes des deux dataframes d'entrée qui ont des valeurs identiques pour certaines colonnes. Ces colonnes sont données par `left_on` et `right_on`.  
    
   
   <div class="alert alert-success" style="border-left: 5px solid #3c763d;border-radius:2px;">Afficher les 5 premieres lignes.</div>



```python
DF = series.merge(users, left_on="Netflix", right_on="Netflix")
DF[:5]
```

Cela nous permet de savoir par série quelles sont les personnes qui peuvent la regarder sur netflix.

<div class="alert alert-success" style="border-left: 5px solid #3c763d;border-radius:2px;">Afficher les noms des 3 personnes les plus agées.</div>

```python
users.sort_values(by='age', ascending=False)[['Nom','age']][0:3]
```
