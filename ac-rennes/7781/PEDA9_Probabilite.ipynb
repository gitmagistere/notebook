## <span style="color:orange"> B. Compter automatiquement les résultats d'une série d'expériences aléatoire ou d'un échantillon.
    
> ### 1. Compter automatiquement les résultats : On reprend le programme "pile ou face"
>> - **On compte les *pile* :** avec une condition "if" 
>> - **Modifiez le programme** pour compter les *[face](https://)*

```python
from random import*
piece = ["pile", "face"]
c = 0    # On crée une variable compteur

for n in range (10) :
  resultat = choice(piece) 
  print (resultat)
  if resultat == "pile" :
    c = c + 1

print (c,"fois pile")
```

>> - **Pour aller plus loin :** on crée deux variables c et d pour compter les *pile* et les *face* sur 100 lancers (on n'affiche plus les résultats)

```python
from random import*
piece = ["pile", "face"]
c = 0    # variable compteur pour les pile
d = 0    # variable compteur pour les face

for n in range (100) :
  resultat = choice(piece) 
  if resultat == "pile" :
    c = c + 1
  else :
    d = d + 1

print (c,"fois pile")
print (d,"fois face")
```

> ### 2. Utilisation de listes
>> - **listes classiques :** comment créer un échantillon de 10 piles ou faces ?

```python
from random import*
piece = ["pile", "face"]
echantillon =[]

for n in range(10) :
    echantillon.append(choice(piece))

n = echantillon.count("face")
print(echantillon)
print("nbre de face =",n)
```

>> - **listes par compréhension :** comment créer un échantillon de 10 piles ou faces ?

```python
from random import*
piece = ["pile", "face"]
echantillon =[choice(piece) for n in range(10)]
 
print(echantillon)
print("nbre de face =",n)
```

## <span style="color:orange"> C. Calcul des fréquences dans un échantillon
>> - **Calculez** la fréquence d'apparition des *pile* sur 50 lancers (On n'affiche plus les résultats pour gagner de la place)
>> - **Modifiez** ce programme pour calculer la fréquence de *face*
>> - **Vous êtes très fort !!!**  créez de nouvelles variables pour afficher les deux fréquences des *pile* et des *face*

```python
from random import*
piece = ["pile", "face"]
c = 0    # On crée une variable compteur

for n in range (50) :
  resultat = choice(piece) 
  if resultat == "pile" :
    c = c + 1

print (c,"fois pile")

f = c/50 # Cette variable est pour la fréquence
print ("frequence de pile =",f)
```
