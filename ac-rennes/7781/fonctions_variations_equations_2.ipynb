#### **1. Découverte des fonctions numeriques et des fonctions python**


- **Une fonction en mathématiques** est un procédé qui transforme un nombre et lui attribue au plus un autre nombre que l'on appelle son image par la fonction.

On note généralement une fonction avec la lettre "f". L'expression de la fonction décrit les opérations à réaliser sur la variable x.

Voici par exemple l'expression de la fonction que nous allons étudier :
\begin{equation*}
f(x) = 0,4x^{2} - x + 2
\end{equation*}


---



- **Fonction python :** Afin d’automatiser les calculs, vous allez programmer cette fonction en langage python et utiliser justement des « fonctions python ». 

Une fonction en langage python est une partie d’un programme que l’on « appelle » quand c’est utile et qui comporte un plusieurs paramètres.
Une fonction python est utile par exemple dans une boucle

```python
# La fonction python appelée "f" a pour paramètre "x" et renvoie son image
def f(x):
	return 0.4*x**2 - x + 1

# La partie principale du programme demande d'entrer "x"
x = float(input("entrez la valeur de x "))
print ("f(",x,")=",f(x))        # affiche la valeur renvoyée par la fonction
```

- Pour calculer automatiquement un ensemble de valeurs sur un intervalle, il nous faut faire des boucles.

La boucle "for" opère sur des itérables tels que les listes, mais généralement sur des listes d'entiers. Pas pratique pour les fonctions...

La boucle "while" permet d'utiliser une variable x que l'on doit initialiser et ensuite incrémenter dans la boucle.

```python
def f(x):
	return 0.4*x**2 - x + 1

for x in range (0,10) :
  print ("f(",x,")=",f(x))       
```

```python
def f(x):
	return 0.4*x**2 - x + 1

x = 0 # On démarre à 0

while x <= 5 :
	print("f(",x,")=",f(x))
	x = x + 0.5
```

Remarque : avec un incrément de 0.1 les floats font des trucs bizarres...
Il est conseillé d'arrondir avec round(f(x),2)

#### **2. Représentation graphique d'une fonction**

- La représentation graphique de la fonction f sur l'intervalle [0 ; 5] est l'ensemble des points de coordonnées (x ; f(x)).
- Tracer la représentation graphique necessite les modules **numpy** et **matplotlib.pyplot**
- Plus le nombre de points est élévé et plus la courbe est "lisse"... on utilise une liste **arange(a,b,pas)** un intervalle [a;b] et un pas

- On peut aussi utliser la fonction **linspace(a,b,n)**... n points dans l'intervalle [a;b]
- Les modules **numpy** et **matplotlib.pyplot** étant importés par les préfixes "np" et "plt" on rajoute "np."devant arange() ou linspace() et "plt." devant plot()

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5,0.1)  # Définit une liste de valeurs de 0 à 5 avec un pas de 0,1
plt.plot(x,0.4*x**2-x+1) # Trace la courbe 

plt.ylabel("axe des ordonnées")
plt.xlabel("axe des abcisses")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,10)  # Définit une liste de 10 valeurs entre 0 et 5
plt.plot(x,0.4*x**2-x+1) # Trace la courbe 

plt.ylabel("axe des ordonnées")
plt.xlabel("axe des abcisses")
plt.grid()
plt.show()
```

#### **3. Variations de la fonction**

- Fonction croissante : Une fonction « f » est croissante sur un intervalle I si quelles que soient les valeurs a et b de cet intervalle :
\begin{equation*}
a < b \Rightarrow f(a) < f(b)
\end{equation*}

- Fonction décroissante : Une fonction « f » est décroissante sur un intervalle I si quelles que soient les valeurs a et b de cet intervalle :
\begin{equation*}
a < b \Rightarrow f(a) > f(b)
\end{equation*}


---



- Analyse du sens de variation de la fonction avec un programme python

L’objectif est de comparer automatiquement les images f(a) et f(b) de deux nombres a et b (séparées par exemple d’un écart e = 0,1). 

Vous choisirez des valeurs a et b de l’intervalle, en partant de 0 et en augmentant a de 0,5 à chaque fois. Le programme affichera « croissant » ou « décroissant ».

```python
def f(x) :
	return 0.4*x**2 - x + 1

a = 0

while a <= 5 :
	b = a + 0.1  # écart entre a et b
	if f(a)<= f(b) : print ("pour a =",a,"la fonction est croissante")
	else : print ("pour a =",a,"la fonction est décroissante")
	a = a + 0.5
```

Remarque : Vous pouvez affiner en modifiant la valeur de l'écart et de l'incrément

#### **4. Maximum ou minimum d'une fonction**

- Maximum d'une fonction sur un intervalle : On appelle maximum de la fonction f sur l’intervalle I, la valeur notée max telle que pour tout nombre x de l’intervalle I on a :
\begin{equation*}
f(x) ≤ max
\end{equation*}

- Minimum d'une fonction sur un intervalle : On appelle minimum de la fonction f sur l’intervalle I, la valeur notée min telle que pour tout nombre x de l’intervalle I on a :
\begin{equation*}
f(x) ≥ min
\end{equation*}


---



- Détection du minimum avec un programme python :

L'hypothèse de départ est bien qu'il y a un minimum dans [0,5].

L'objectif de ce programme python est de détecter automatiquement la valeur du minimum et son abscisse.

On compare donc f(a) et f(b)...à cette fin l'écart entre a et b est fixé 0,01 pour obtenir la valeur du mininmum avec cette précision.

La valeur de l'abscisse du minimum est arbitrairement choisie comme étant la valeur de a lorsque f(a) > f(b)

L'incrément est également de 0,01

```python
def f(x) :
	return 0.4*x**2 - x + 1

a = 3

while a <= 5 :
	b = a + 0.01
	if f(a)<= f(b) : break
	a = a + 0.01

print ("la valeur minimale est",round(f(a),4))
print ("elle est obtenue pour x =",round(a,4))
```

#### **5. Résolution d'équation du type f(x) = a**

- Résolvons par exemple l'équation suivante :
\begin{equation*}
f(x) = 7
\end{equation*}

L'hypothèse de départ est que la solution se trouve dans l'intervalle [3,4] et on transforme la fonction pour déterminer la solution de :

\begin{equation*}
f(x) - 7 = 0
\end{equation*}

- La méthode de résolution par dichotomie peut être traduite par cet algorithme en langage naturel :

algo1.png

```python
def f(x) :
	return 0.4*x**2 - x + 1 - 3   # la fonction moins 7

a = 3 # intervalle de départ
b = 4

while abs(b-a) > 0.01 :     # choix de la précision
  m = (a+b)/2
  if f(a)*f(m) < 0 :
    b = m
  else :
    a = m

a = round(a,3)
b = round(b,3)
print ("la solution est entre ",a," et ",b)
```
