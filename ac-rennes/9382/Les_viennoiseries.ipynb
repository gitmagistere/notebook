# Avant de commencer...
</div>
    


Cet environnement de travail s'appelle un *notebook Jupyter* : à l'intérieur d'un navigateur Web, alternent
- des cellules de média (comme celle que tu es en train de lire) avec du texte, des images, des liens ;
- des cellules de code (destinées à contenir du Python).

Pour exécuter une cellule de code :
- clique sur la cellule pour qu'elle soit encadrée en bleu (selectionnée) ;
- si ton curseur de texte clignote à l'intérieur : tu peux modifier le contenu de la cellule (sinon clique à l'intérieur) ;
- pour exécuter la cellule, il y a le bouton « Exécuter » du menu du haut, ou mieux, le raccourci `MAJ + Entrée`.

Quand une cellule **n'a pas encore exécutée**, dans sa marge gauche on peut lire `IN [ ]` (ou `Entrée [ ]`).

Une fois exécutée, on pourra lire `IN [n]`, où `n` représente le nombre de fois qu'on a exécuté n'importe laquelle des cellules de code.
    
<div style="color:#F4D03F;">
    
### En théorie
</div>
    
Le notebook est écrit pour que les cellules soient exécutées les unes après les autres, dans l'ordre de la lecture : en effet, toutes les exécutions sont « gardées en mémoire ».

    
</div>


### Exemple d'exécutions successives de cellules

On commence par créer une variable `a` valant 2

```python
a = 2 # Merci d'exécuter la cellule. Une fois cela fait tu pourras lire "Entrée [1]" à gauche de celle-ci.
```

```python
print(a)
```

Si tu exécutes plusieurs fois la cellule précédente, le nombre entre crochet augmentera : c'est le total de cellules qui ont été exécutées au cours de la séance.

La variable `a`est persistante : on peut s'en resservir dans d'autres cellules :

```python
print(a) 
a = a + 1
print(a)
```

**La chronologie du notebook, c'est celle de l'éxecution des cellules** : si tu exécutes la deuxième cellule à partir du haut, elle afficher la valeur actuelle de `a`: remonter dans le notebook, ce n'est pas « remonter dans le temps » !
