p5 est une bibliothèque spécialement conçue pour réaliser des graphiques simplement.
Elle suit le *paradigme de programmation événementielle*, c'est-à-dire qu'on écrit des fonctions qui seront appelées automatiquement suivant certains événements.

Nous n'avons pas encore vu la notion de fonction en cours, contentons-nous de dire que ce sont des « petits morceaux de programmes » que l'on peut exécuter autant de fois que nécessaire.

Dans la cellule suivante, on définit les objets que tu vas *indirectement* utiliser :

````python
import p5

SCREEN_WIDTH = 200   # largeur fenêtre graphique
SCREEN_HEIGHT = 200  # hauteur
SIZE = 20            # taille carre

WHITE = (255, 255, 255) # définition des couleurs
RED = (255, 0, 0) #
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)


def setup() :
    p5.createCanvas(SCREEN_WIDTH, SCREEN_HEIGHT)    # Création fenêtre graphique
    p5.background(0)                                # le fond est noir
    p5.noLoop()                                     # draw() ne sera appelée qu'une fois
````

Maintenant, voici *les seuls objets* que tu pourras utiliser :

````python
WHITE = (255, 255, 255) # définition des couleurs
RED = (255, 0, 0) #
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

def square(i,j,col):                  # remplit le carré (i,j) de la couleur souhaitée (WHITE, RED...)
    p5.fill(col)
    p5.rect(i*SIZE,j*SIZE,SIZE,SIZE)
````

Pour utiliser ces objets, c'est très simple : il suffit de 
- définir *le corps* de la fonction `draw`, c'est-à-dire, programmer ce que cette fonction (qui sera appelée automatiquement) doit faire;
- lancer p5 avec l'instruction `p5.run()`.

Dans l'exemple suivant, on décide de colorier le carré $(0,0)$ en bleu:

````python
def draw():           # toujours commencer par cette ligne (draw est une fonction qui sera appelée automatiquement)
    square(0,0,BLUE)  # on va juste dessiner le carré (0,0) en bleu 

p5.run()              # et on « démarre »
````

## Un peu d'animation

Tout au début de ce document on a écrit :
```
def setup() :
    p5.createCanvas(SCREEN_WIDTH, SCREEN_HEIGHT)    # Création fenêtre graphique
    p5.background(0)                                # le fond est noir
    p5.noLoop()                                     # draw() ne sera appelée qu'une fois
```

La dernière ligne entraîne que la fonction draw n'est appelée qu'une seule fois.
Retirons-la :

````python
def setup() :
    p5.createCanvas(SCREEN_WIDTH, SCREEN_HEIGHT)    # Création fenêtre graphique
    p5.background(0)                                # le fond est noir
    p5.frameRate(2)
````

Maintenant la fonction `draw` sera appelée 2 fois par seconde (2 FPS).

On peut également importer la fonction `randint` de la bibliothèque `random`. C'est une fonction qui

- en entrée prend 2 `int` a et b;
- renvoie un entier choisi « au hasard » entre a et b compris (pas comme un `range`, donc).

````python
from random import randint

print(randint(0,255))
````
