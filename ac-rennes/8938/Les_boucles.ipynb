# **Boucles for et while**


### **1. Utilisation des boucles en algoritmique**


---





Les structures itératives permettent d’exécuter plusieurs fois un bloc d’opérations.

- **Boucles bornées (répéter n fois) :** La boucle « FOR » permet de répéter un bloc d’opérations un nombre de fois défini en faisant varier automatiquement la valeur d’une variable (compteur ou variable de boucle).



*   La variable "compteur" est créée automatiquement, elle n'a pas besoin d'être initialisée et incrémentée. Le pas est 1 par défaut dans python
*   Il n'y a pas dans le langage python de boucle "répéter" contrairement à scratch, mais c'est l'occasion de rappeler la notion d'intervalle.



1.png



- **Boucles non bornée :** « WHILE » permet de répéter un bloc d’opérations **tant qu’une condition** est satisfaite. Il faut être attentif au problème de la terminaison et ne pas créer de boucle infinie.

### **2. Introduction progressive des boucles : la boucle FOR**


---




- **la variable compteur** prend toutes les valeurs d'une liste ou d'une autre variable. 


*   La boucle commence par FOR et termine par deux points.
*   Toutes les instructions indentées (décalées) font partie de la boucle.
*   La variable "compteur" n'est pas forcément utilisée








```python
for i in "crocodile" :
  print(i)
```

```python
for x in range(4) :  # l'instruction range(n) crée une liste de valeurs de 0 à n-1
  print (x)
```

```python
for x in range (0,4) : # c'est la même chose mais c'est plus explicite cela correspond à l'intervalle [0;4[
  print (x)
```


*   for c in .....:crée **une variable compteur**
*   L'instruction **range(a,b)** crée une liste de valeurs entières entre a et b-1
*   L'instruction **range(a)** crée une liste de valeurs entières entre 0 et a-1
*   On peut tout de même avoir un pas différent de 1 



```python
for x in range (2,12,2) :  # on peut avoir un pas entier, mais pas décimal , sinon voir boucle while
  print(x)
```

- **Exemple d'utilisation mathématique :** calculer l'image de x par une fonction sur l'intervalle [0;5] ou pour les suites 

```python
for x in range (0,5) :
  print (x,"-->",x**2) 
```

```python
u = float(input("premier terme U1=")) # premier terme de la suite
r = float(input("raison r=")) 

for n in range (1,6) : # ici n ne sert à rien dans les calculs
  print("U",n,"=",u)
  u = u + r  # la position de cette ligne n'est pas anodine
```

```python
from math import * # importation du module math pour utiliser la racine carrée
for x in range (0,6) :
  y = round (sqrt(x),2)  # arrondi à deux décimales 
  print (x,"a pour image",y)
```

- Il peut être frustrant à ce stade de ne pas pouvoir utiliser de valeurs décimales pour la variable. Mais il faut alors utiliser la boucle **while**

### **3. La boucle WHILE**


---






*   A la différence de la précédente elle n'a pas de borne et peut être exécutée indéfiniment
*   Attention !!! si vous executez le code ci-dessous il faudra mettre en pause très rapidement !!!! aussi je n'ai pas écrit les deux points après True... faites le test dans votre interprèteur python....



```python
while True # erreur volontaire : cette boucle est infinie
  print("OK")
```



- Vous devez créer et initialiser la variable compteur avant la boucle
- Vous devez penser à incrémenter cette variable à la fin des instructions de la boucle, ou au début... au choix
- C'est l'occasion de rappeler les inégalités, les encadrements...
- Les opérateurs de comparaison sont : ==; !=, <= etc
- La condition peut être un encadrement a < x < b


```python
x = 0  # on crée et initialise la variable de la boucle

while x < 5:
  print (x,"-->",x**2) 


# ça ne marche pas ? devinez pourquoi...
# !!!! mais surtout stoppez l'exécution de la cellule avant !!!
```

```python
x = 0  # on crée et initialise la variable de la boucle

while x < 2:
  print (x,"-->",x**2) 
  x = x + 0.5
  # si vous changez l'incrément par 0.1 vous allez avoir des surprises...
```

- Les nombres à virgules ont quelques limitations dans python voir sur le lien suivant :
[problèmes et limites](https://docs.python.org/fr/3/tutorial/floatingpoint.html)

```python
x = 0

while x < 2 :
  print (round(x,2),"-->",round(x**2,2)) 
  x = x + 0.1
```

- Il peut être interessant de fixer un objectif à une suite numérique :


```python
u = 3 # premier terme de la suite
r = 5.5 # raison
n = 1 # indice de la suite
while u < 90 : 
  print("U",n,"=",u)
  u = u + r  
  n = n + 1
```

- Résolution par balayage d'une inéquation de type : $$f(x)<a$$



```python
x = 0
y = 0 # il faut la créer même si la valeur initiale ne sert à rien

while y <= 5 :
  y = x**2
  print (x,"-->",y) 
  x = x + 0.5
  # Modifiez l'incrément pour affiner la résolution
  # remarque : on dépasse l'objectif... pourquoi ?
```

- Deux possibilités pour ne pas dépasser l'objectif :



```python
x = 0
y = 0 # il faut la créer même si la valeur initiale ne sert à rien

while y <= 5 :
  print (x,"-->",y) 
  x = x + 0.5
  y = x**2
# il nous manque quand même une valeur au départ
```

```python
x = 0
y = 0 # il faut la créer même si la valeur initiale ne sert à rien

while y <= 5 :
  y = x**2
  if y > 5 : break # on peut aussi résoudre en partie la problèmatique précédente avec l'instruction break
  print (x,"-->",y) 
  x = x + 0.5
```
