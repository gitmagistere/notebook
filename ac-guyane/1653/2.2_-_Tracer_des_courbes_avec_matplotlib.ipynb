# Tracer des courbes avec matplotlib

La librairie permettant de tracer les graphiques avec python est **matplotlib**.

Site officiel : https://matplotlib.org/

Les tracés se font avec un module de matplotlib appelée **pyplot**.

En pratique, on utilisera donc les fonctions contenues dans `matplotlib.pyplot`

> Remarque : matplotlib est déjà disponible dans Edupython, vous pouvez donc directement importer la librairie. Sur un notebook par contre, il faudra ajouter une ligne en plus que vous verrez plus tard.

## Appel du module et définition d'un alias

Pour appeler le module pyplot et pouvoir accéder à ses fonctions, on ajoutera la ligne suivante en début de code :

```python
import matplotlib.pyplot as plt
```

On importe le module, et afin de ne pas avoir à taper à chaque fois "matplotlib.pyplot" on lui donne un alias "plt"
Ainsi par la suite, une fonction "show" contenue dans le module "pyplot" de la librairie "matplotlib" pourra être appelée en tappant :
```python
plt.show()
```
Plutôt que de tapper : 
```python
matplotlib.pyplot.show()
```

### Graphique 2 

Enrichissons avec quelques "décorations"

````python
plt.plot([1,2,3,4],[10, 20, 30, 40],'r+:',label = 'une courbe au hasard')
plt.ylabel('label des y')
plt.xlabel('label des x')
plt.title('Un graphique très peu utile')
plt.legend()
plt.grid()
plt.show()
````

La décoration a été obtenue en ajoutant différentes options. Cette image illustre les différentes parties d'un graphique : 
![caption](https://physique-chimie.dis.ac-guyane.fr/sites/physique-chimie.dis.ac-guyane.fr/IMG/png/decoration.png)

### Pour résumer, pour obtenir un graphique on va :
1. Préparer les données (x, y) et les indiquer dans une courbe : plt.plot
> Remarque : Un graphique peut convenir plusieurs courbes
2. Configurer le graphique : couleur, ajout de légende, labels, histogramme, .. en ajoutant des options au graphique
3. Demander à afficher le graphique par plt.show()
