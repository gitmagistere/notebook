### Les boucles non-bornées

<p style="color:green;font-size:1.3em;"> <u><b>Définition</b></u> :
Dans un algorithme, une boucle est une suite d'instructions que l'on répète (en boucle) un certain nombre de fois. On dit que la boucle est <b>non bornée</b> lorsque l'on ne sait pas <i>à l'avance</i> le nombre d'itérations nécessaire pour effectuer la boucle.
</p>

Malheureusement, la plupart du temps, on ne sait pas combien de fois on va devoir recommencer une opération... Par exemple, compter le nombre de chiffre dans $2022^{2022}$ : si on avait déjà la réponse, il n'y aurait pas besoin de compter !   
Il nous faut donc une structure *tant que* qui nous permet de continuer à répéter une suite d'instruction jusqu'à ce qu'une certaine condition **ne se réalise plus**. Il faudra donc faire attention à être sûr que la condition puisse à un moment ne plus se réaliser, sinon, on va créer une boucle infinie : le programme va continuer à fonctionner sans jamais s'arrêter. 

````python
from tutor import tutor
def nbre_chiffres(n):
    "Détermine le nombre de chiffres de l'entier n"
    i = 0
    while n // (10**i) !=0:
        i=i+1
    return i

nbre_chiffres(173897)
tutor()
````

Une population de coccinelles compte initialement  200 individus. Le modèle mathématique d'étude utilisé, en tenant compte de différentes contraintes biologiques (décès, naissances, insecticides, pesticides, prédateurs féroces, enfants curieux..) est le suivant. On estime que la population $U_{n+1}$ de l'année $n+1$ se déduit de la population $U_n$ de l'année $n$ par la relation : $U_{n+1} = 2,8 U_n - 0,002 U_n^2$. On pose $U_0$ = 200. Déterminer au bout de combien d'années la population dépasse les 800 individus.
!!!

!!! success La réponse
<details>

```python
def population(seuil):
    "Détermine l'année à laquelle la population de coccinelle dépasse un seuil"
    u = 200
    n = 0
    while u < seuil:
        u = 2.8*u-0.002*u**2
        n = n + 1
    return n

population(800)
```
