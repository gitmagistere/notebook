# <p style="text-align:center;"> Formation Python Débutant - Les fonctions en Python</p>

La syntaxe de la déclaration d'une fonction est la suivante : 
```
def nom_de_la_fonction(parametre_1, parametre_2, paramètre_3, ...):
    "Commentaire sur la fonction"
    Ce que fait la fonction
    return resultat_1, resultat_2, resultat_3, ...
```


Pour définir une fonction il faut : 
- utiliser le mot clé `def` : c'est ce mot clé qui permet de déclarer la création d'une fonction. Ainsi `def` ne peut pas être utilisé pour déclarer une variable ;  
- donner un nom à votre fonction : il doit respecter les mêmes règles de définition que les noms de variables ;   
- définir les paramètres de votre fonction : il peut y en avoir entre 0 et n ;  
- finir la première ligne par le caractère `:` ;  
- décrire ce que fait votre fonction. C'est un texte sommaire encadré par des guillemets s'il est écrit sur une ligne et par 3 guillemets s'il est écrit sur plusieurs lignes. L'objectif de ce commentaire n'est pas d'indiquer le procédé, qui sera décrit par le corps de la fonction, mais de préciser éventuellement les différentes valeurs possibles pour les paramètres et les résultats renvoyés par la fonction. Cette description peut être retrouvé en appelant l'aide de la fonction : cela se fait avec la fonction `help`. 
- le corps de la fonction est **indenté** : tout ce qui se passe dans la fonction est décalé vers la droite d'une tabulation (ou de 4 espaces) et n'est pas aligné à gauche. L'indentation se fait avec votre touche de tabulation (à gauche de la lettre "a" sur le clavier). Si l'on revient aligné à gauche, on sort automatiquement de la fonction ; 
- La fonction se termine par le mot clé `return` qui indique ce que renvoie la fonction : un ou plusieurs résultats.

!!! warning **Pour éviter quelques bugs liés aux navigateurs, n'exécuter pas la cellule suivante, le résultat est déjà affiché**
!!!

````python
def f(x):
    "Fonction carrée. x est un réel quelconque. La fonction renvoie le carré de x."
    return x**2

help(f)
print("Le carré de 3 est ",f(3))
````

Nous avons vu au passage le mot clé Python `help` : 
- qui permet d'obtenir de l'aide sur une fonction que nous avons créée (ou que quelqu'un d'autre a créée) : cela permet de savoir ce que fait la fonction et quelles sont ses paramètres ; 
- que l'on peut aussi appliquer aux fonctions Python déjà existantes pour avoir des informations supplémentaires. Elle peut notamment s'appeler elle même.

!!! warning **Pour éviter quelques bugs liés aux navigateurs, n'exécuter pas la cellule suivante, le résultat est déjà affiché**
!!!

````python
help(help)
````

Dans la définition de la fonction, on peut décider de la valeur *par défaut* d'un paramètre de la fonction : il n'est alors pas nécessaire pour l'utilisateur de donner une valeur à ce paramètre si la valeur par défaut lui convient.

````python
def somme(a,b=1):
    "Fonction somme des nombres a et b. Par défaut, la valeur de b est 1."
    return a + b

# print sert à avoir un affichage "joli" pour comprendre comment marche Python.

print("La somme de 2 et de 1, valeur par défaut de b vaut", somme(2))
print("la somme de 2 et de 5 vaut ",somme(2,5))
````

````python
# On aurait aussi pu faire deux entrées en console :
somme(2)
````

````python
somme(2,5)
````

Un autre exemple : 

````python
def plus_grand_multiple_de_b_inférieur_à_a(a,b):
    "Fonction qui renvoie le plus grand multiple de b inférieur ou égal à a"
    return b * (a//b)

# print sert à avoir un affichage "joli" pour comprendre comment marche Python.

print("Le plus grand multiple de 6 inférieur à 100 est ", plus_grand_multiple_de_b_inférieur_à_a(100,6))
````
