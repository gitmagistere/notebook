### Les boucles bornées

<p style="color:green;font-size:1.3em;"> <u><b>Définition</b></u> :  
    En <i>informatique</i>, une <b>boucle bornée</b> est une struture qui permet de recommencer une procédure un nombre connu de fois. 
</p>

Regardons tout de suite un exemple : 

````python
for i in range(5):
    print("bonjour")
````

Regardons de plus près : 
- le mot clé Python permettant la répétition est `for` 
- j'ai besoin d'un indice de répétition, souvent en mathématique c'est `i`. S'il ne sert à rien, on peut le remplacer par `_`.
- je dois décrire dans quel ensemble est `i` à l'aide du mot clé `in`.
- dans le cas d'un ensemble de 5 nombres, j'utilise le mot clé `range` .
- je termine ma ligne par `:`.
- ce qui doit être fait dans ma boucle est indentée.

Revenons plus en détail sur le mot clé `range` qui peut s'écrire de différentes manières : 
- `range(a)` : décrit les entiers de 0 à `a-1` (`a` est exclu), il y a donc a nombres
- `range (a,b)` : décrit les entiers de `a` à `b-1` (`b` est exclu), il y a donc b-a nombres
- `range(a,b,c)` : décrit les entiers de `a` à `b-1` (b est exclu) avec un pas de c (c'est-à-dire les nombres `a`, `a+c`, `a+2c`, ...), il y a donc (b-a)//c + 1 nombres

En général on utilise les deux premiers types de commande : 

````python
for i in range(5,10):
    print(i)
````

***Un exemple simple :***   
Calculer la somme des entiers naturels de 0 à $n$. On va réaliser le calcul comme on le ferait "à la main" : on part de 0, on rajoute 1, puis on rajoute 2 au résultat, puis on rajoute 3 au résultat, puis on rajoute 4 au résultat, puis.... 

````python
def somme_entiers(n):
    "Calcul la somme des entiers de 0 à n-1"
    S = 0
    for i in range(n):
        S = S + i
    return S

somme_entiers(10)
````

!!! example <p style="color:red;"> A vous de faire : </p>
Une population de coccinelles compte initialement  200 individus. Le modèle mathématique d'étude utilisé, en tenant compte de différentes contraintes biologiques (décès, naissances, insecticides, pesticides, prédateurs féroces, enfants curieux..) est le suivant. On estime que la population $U_{n+1}$ de l'année $n+1$ se déduit de la population $U_n$ de l'année $n$ par la relation : $U_{n+1} = 2,8 U_n - 0,002 U_n^2$. On pose $U_0$ = 200. Déterminer l'effectif de la population de coccinelles  au bout de 10 ans puis de 20 ans.

*Cet exercice permet d'introduire la fonction `round(a,b)` qui arrondit le nombre `a` avec `b` chiffres après la virgule.*
!!!

!!! success La réponse
<details>
    <summary style="color:red;">&rarr;Cliquez-ici pour la réponse&larr; </summary>
    
```python
def pop_coccinelles(n):
    "Détermine la population de coccinelles au bout de n années."
    u = 200
    for _ in range(n):
        u = 2.8*u-0.002*(u**2)
    return round(u,0)

pop_coccinelles(20)
    
```

La même somme est placée au même moment à 1,5% à intérêt composé. De quelle somme dispose-t-on au bout de 10 ans ?  
!!!

!!! success La réponse
<details>
    <summary style="color:red;">&rarr;Cliquez-ici pour la réponse&larr; </summary>
    
```python
def interet_simple(somme, interet, annee):
    capital = somme
    for i in range(annee):
        capital = capital + somme*interet
    return capital

def interet_compose(somme, interet, annee):
    capital = somme
    for i in range(annee):
        capital = capital*(1+interet)
    return capital

    
# J'utilise print pour avoir deux réponses sans avoir à faire les appels dans la console    
print(interet_simple(2500,1.6/100,10))
print(round(interet_compose(2500,1.5/100,10),2))
```
