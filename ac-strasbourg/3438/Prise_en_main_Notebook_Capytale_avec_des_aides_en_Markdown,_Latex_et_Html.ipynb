# II. $\LaTeX$ <a name="bas"></a> 

Pour que le code s'affiche, il faut le mettre entre deux dollars : <code>$ ... $</code>  
Pour écrire la formule en mode maths, centrée sur une ligne, il faut doubler les dollars : <code>$$ ... $$</code> voir [fractions](#frac) pour comprendre.  
On trouvera une liste plus complète des commandes Latex en cliquant <a href="https://fr.wikibooks.org/wiki/LaTeX/%C3%89crire_des_math%C3%A9matiques ">ici</a> mais ce qui suit devrait largement suffire!
  
## 1. Bases 

### a. multiplication <a name="mul"></a> 
On utilise <code>\times</code> pour obtenir le symbole « multiplié par » $\times$.  
Exemple : <code>2\times3=6</code> donne $2\times3=6$. 
### b. racines <a name="rac"></a>  
<code>\sqrt{a+b}</code> pour écrire $\sqrt{a+b}$, mais on peut se contenter de <code>\sqrt a</code> pour écrire $\sqrt a$. 

Pour les racines nième, on écrira <code>\sqrt[n]{}</code>. Par exemple, <code>\sqrt[3]{8}</code> pour $\sqrt[3]{8}$.

### c. puissances<a name="pui"></a>  
<code>x^{2n-3}</code> pour écrire $x^{2n-3}$ , mais on peut se contenter de <code>x^n</code> pour écrire $x^n$.  


### d. fractions <a name="frac"></a>
On utilise <code>\frac{}{}</code> pour obtenir le trait de fraction.  
Exemple : on peut taper <code>\frac{x+2}{2x-3}</code> pour écrire $\frac{x+2}{2x-3}$, mais on peut se contenter de  <code>\frac ab </code>pour écrire $\frac ab $.  
<b>Remarques :</b> on pourra utiliser <code>\tfrac</code> pour avoir des fractions au sein du texte qui ne modifient pas l'interligne (exemple : un « petit » $\tfrac{1}{2}$), et de <code>\dfrac</code> pour avoir des grosses fractions.
Il est d'ailleurs parfois utile de combiner les deux :
$$\dfrac{1 + \tfrac{1}{2}}{3} \text{ est plus joli que } \dfrac{1 + \dfrac{1}{2}}{3}$$
<code>\frac</code> a l'avantage de s'adapter au contexte : il vaut <code>\tfrac</code> dans le corps du texte comme ici $\frac{0}{0}$ et <code>\dfrac</code> en mode maths centré : $$\pi \approx \frac{22}{7}$$.

### e. parenthèses de bonne taille <a name="par"></a>
Écrire <code>\left(</code> et <code>\right)</code> pour que la parenthèse prenne la bonne taille.  
Exemple : <code>5(2 +\frac{\sqrt x}{2})</code> donne $5(2 +\frac{\sqrt x}{2})$, tandis que <code>5\left(2 +\frac{\sqrt x}{2}\right)</code> donne $5\left(2 +\frac{\sqrt x}{2}\right)$  
<code>\left</code> et <code>\right</code> s'applique à toutes sortes de parenthèses : <code>\left[</code> et <code>\right]</code>, <code>\left\{</code> et <code>\right\}</code>, <code>\left|</code> ou <code>\left\vert</code> pour les valeurs absolues |, <code>\left\Vert</code> ou <code>\left\\|</code> pour les normes $\left\|u\right\|$.

## 3. Matrice et système <a name="mat"></a>  

### a. coordonnées
On peut utiliser les matrices pour écrire les coordonnées en colonne :\begin{pmatrix}   x_\overrightarrow{AB} \\  y_\overrightarrow{AB}\end{pmatrix}  
On l'obtient en tapant <code>\begin{pmatrix}   x_\overrightarrow{AB} \\  y_\overrightarrow{AB}\end{pmatrix}</code> 
### b. systèmes d'équations
$$\left\{\begin{align*} x + y &= 2 \\ 3x - 2y &= -5 \end{align*}\right.$$
On l'obtient en tapant <code>\left\\{\begin{align*} x + y &= 2 \\ 3x - 2y &= -5 \end{align*}\right.</code>  
Remarque : avec<code>align*</code> le code <code>&</code> définit sur quoi aligner.  
### c. pour des égalités bien alignées <a name="aut"></a> 
$$\begin{align*} 
A &= \sqrt{144} & \text{d'après truc} \\ 
&= 12 & \text{car je sais calculer} \\
&= 2^2 \times 3 &\text{décompo facteurs premiers}
\end{align*}$$
On l'obtient en tapant <code>\begin{align*} 
A &= \sqrt{144} & \text{d'après truc} \\ 
&= 12 & \text{car je sais calculer} \\
&= 2^2 \times 3 &\text{décompo facteurs premiers}
\end{align*}</code>  
### d. aligner après un espace
On utilise <code>cases</code> pour des cas où on aligne après un espace, parfait pour des conditions :
$$|x| = \begin{cases} x & \text{si } x \geqslant 0 \\ -x & \text{si } x < 0\end{cases}.$$ 
Obtenu en faisant <code>|x| = \begin{cases} x & \text{si } x \geqslant 0 \\ -x & \text{si } x < 0\end{cases}</code>  
   
   

 ### e. coordonnées dans l'espace
 Coordonnées sur trois lignes : $\begin{pmatrix}   1 \\  2\\  3 \end{pmatrix}$ en tapant  <code>\begin{pmatrix}   1 \\  2\\  3 \end{pmatrix}</code>  
 ### f. de vraies matrices
 Matrice 2*2 : $\begin{pmatrix}   a_1 & b_1 \\  a_2 & b_2 \end{pmatrix}$ en tapant <code>\begin{pmatrix}   a_1 & b_1 \\  a_2 & b_2 \end{pmatrix}</code>

## 6. Géométrie <a name="vec"></a> 
### a. vecteurs  
On tape  <code>\overrightarrow{AB}</code>, <code>\vec{u}</code> pour obtenir $\overrightarrow{AB}$, $\vec{u}$.

Le premier code ajuste la longueur de la flèche au nom du vecteur : par exemple <code>\vec{AB}</code> donne $\vec{AB}$ (c'est moche !) mais <code>\overrightarrow{u}</code> donne $\overrightarrow{u}$.

### b. coordonnées 
Pour les coordonnées de vecteurs, le plus simple pour écrire les coordonnées en colonne c'est :  
<code>\binom{x_M-2}{y_M+3}</code> qui donne $\binom{x_M-2}{y_M+3}$    
mais on peut aussi utiliser les [matrices](#mat)  

Comparaison d’affichage :
* sans matrice : $\overrightarrow{AM}\binom{x_M-2}{y_M+3}$ 
* avec matrice : $\overrightarrow{AM}\begin{pmatrix}   x_M-2 \\  y_M+3\end{pmatrix} $

### c. angles 
On tape respectivement <code>\widehat{(\overrightarrow{AB},\overrightarrow{AC})}</code> pour obtenir $\widehat{(\overrightarrow{AB},\overrightarrow{AC})}$  
et <code>\widehat{ABC}</code> pour obtenir $\widehat{ABC}$

### d. normes    
Plusieurs façons d’écrire :
* <code>\Vert\overrightarrow{AB}\Vert</code> $\Vert\overrightarrow{AB}\Vert$

* <code>\left\Vert\overrightarrow{AB}\right\Vert</code> $\left\Vert\overrightarrow{AB}\right\Vert$

* <code>\\|\vec{u}\\|</code> $\|\vec{u}\|$

### e. autre 
On tape respectivement <code>\overline{AB}</code>, <code>\bar{a}</code> pour obtenir $\overline{AB}$, $\bar{a}$

On tape respectivement <code>\perp,//,\cdot </code> pour obtenir $\perp,//,\cdot $  

Repère :   
`\left(O; \vec{u}, \vec{v} \right)`  pour écrire $\left(O;\vec{u},\vec{v} \right)$

avec i et j, c'est pas beau d'après Alex, parce que le point et la flèche se « téléscopent ».

`\left(O;\vec{\imath},\vec{\jmath} \, \right)` pour écrire  $\left(O ;\vec{\imath},\vec{\jmath} \, \right)$

## 9. Encadrer une formule, écrire du texte, polices  <a name="font"></a>  
### a. encadrer  

Pour encadrer une formule on utilise <code>\boxed{}</code>  
Exemple: <code>\boxed{(x-3)^2 =x^2-6x+9}</code> donne $\boxed{(x-3)^2 =x^2-6x+9}$  

Pour encadrer du texte on utilise <code>\fbox</code>  
Exemple: <code>\fbox{C'est pas beau ?}</code> donne $\fbox{C'est pas beau ?}$  

On peut même encadrer plusieurs fois!  
$\boxed{\text{pour tout réel }x, \boxed{\boxed{(x-3)^2 =x^2-6x+9}}}$

### b. Écrire du texte au milieu d'une formule  
Pour écrire du texte au milieu d'une formule on utilise <code>\text{}</code>

Exemple : <code>\dfrac{\text{ Numérateur }}{\text{ Dénominateur }}</code> donne $\dfrac{\mbox{ Numérateur }}{\mbox{ Dénominateur }}$

### c. Différentes polices à utiliser  


**Pour les ensembles de nombres,** <code>\mathbb</code> ou <code>\Bbb</code> : $\Bbb{CHNQRZ}$.  
même si on devrait plutôt utiliser <code>\mathbf</code> :$\mathbf{ABCDEFGHIJKLmnopqrstuvwxyz}$.  
**Pour les ensembles de points,** <code>\mathcal</code> : $\mathcal{ABCDEFGHIJKLmnopqrstuvwxyz}$.  
Pour des lettres bien joiles <code>\mathscr</code> : $\mathscr{ABCDEFGHIJKLmnopqrstuvwxyz}$.

## F. Fonctions python <a name="6"></a>

Comme dans un éditeur tel Pyzo, on peut définir des **fonctions** et les utiliser. En particulier, on peut en créer une dans une cellule et l'utiliser partout ailleurs dans le notebook.

````python
def somme(L):
    """ Renvoie la somme des éléments de la liste L """
    s = 0
    for x in L:
        s = s + x
    return s
````

**11.** Exécuter la cellule suivante pour calculer la somme des termes de la liste.

````python
def seuil(val:float) -> int:
    u = -2
    i = 0
    while u < val:
        i = i + 1
        u = 3*i - 2
    return i
````
