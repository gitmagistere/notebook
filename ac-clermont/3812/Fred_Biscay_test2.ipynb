# Découvrir la loi des grands nombres (partie 2)

Dans le programme précédent, vous avez testé l'influence de la taille de l'échantillon $n$ sur la fréquence $f$ calculée. Afin de bien visualiser l'évolution des fréquences quand $n$ augmente, exécutez le programme suivant. Que constatez-vous ?

```python
import random
import matplotlib.pyplot as plt

nb_de_lancers=10
X=[]
Y=[]

for i in range (100):
    compteur=0

    for j in range (nb_de_lancers):

        resultat_du_de=int(random.random()*6+1)

        if resultat_du_de==6:
                compteur=compteur+1
                
    frequence=compteur/nb_de_lancers
    X.append(nb_de_lancers)
    Y.append(frequence)
    nb_de_lancers=nb_de_lancers+100
    
plt.scatter(X,Y,c="red")
plt.show()
```
