```python
coups_possibles=[]
coups_init=[]
def Effacer():
    """ Efface la zone graphique """
    global plateau
    plateau = [['-'] * 8 for i in range(nb_cases+1)]

def afficher_plateau():
    """ Dessine la grille """
    global plateau
    Effacer()
    plateau[positionJ1//50][2]='O'
    plateau[positionJ2//50][5]='X'
    for i in range(nb_cases+1):
        for j in range(8):
            print(plateau[i][j],end='')
        print()
    for i in range (len(sequence_en_cours)):
        print(sequence_en_cours[i],end='')
    print()

def init_jeu(joueur_deb,nouvelle):
    global coups_possibles,positionJ1,positionJ2,joueur,croisement,distance_totale,nb_cartes,plateau,fin_partie,nb_cases,sequence_en_cours,coups_init
    joueur=joueur_deb
    croisement=False
    if nouvelle:
        distance_totale=int(input("Donner la distance a parcourir par tranche de 50 km comprise entre 50 à 1000 km"))
        while(distance_totale%50!=0 or distance_totale>1000 or distance_totale<50):
            distance_totale=int(input("ATTENTION : Donner la distance a parcourir par tranche de 50 km comprise entre 50 à 1000 km"))
        nb_cartes=int(input("Donner le nombre de cartes disponibles entre 2 et 10"))
        while(nb_cartes<2 or nb_cartes>10):
                nb_cartes=int(input("ATTENTION : Donner le nombre de cartes disponibles entre 2 et 10"))
        for i in range (nb_cartes):
            carte=int(input("Donner la valeur de la carte multiple de 50 km comprise entre 50 à 300 km"))
            while(carte%50!=0 or carte<50 or carte>300):
                carte=int(input("ATTENTION: Donner la valeur de la carte multiple de 50 km comprise entre 50 à 300 km"))
            coups_possibles.append(carte)
        coups_init=coups_possibles[:]#le : pourque lorsque l'on modifie l'un l'autre ne le soit pas
        print(coups_init)
    else:
        coups_possibles=coups_init[:]
    positionJ1=0
    positionJ2=distance_totale
    nb_cases=int(distance_totale/50)
    plateau = [['-'] * 8] * nb_cases
    sequence_en_cours=[]
    fin_partie=False

def coups_encore_possible(joueur):
    for e in coups_possibles:
        if joueur==1 and (positionJ1+e)<=distance_totale:
            return True
        elif joueur==2 and (positionJ2-e)>=0:
            return True
    return False

def jouer(valeur):
    global coups_possibles,positionJ1,positionJ2,joueur,croisement,fin_partie,sequence_en_cours
    if joueur==1:
        if positionJ1+valeur>=positionJ2 and not croisement:
            positionJ1=positionJ2
            croisement=True
            sequence_en_cours.append("J1_"+str(coups_possibles.pop(coups_possibles.index(valeur)))+"  ")#On enleve de la liste le coup que l'on vient de jouer
            if positionJ2!=0:
                joueur=2
        else:
            if positionJ1+valeur>distance_totale:
                print("ATTENTION:Vous ne pouvez pas jouer ce coup car vous sortez du plateau")
                if not coups_encore_possible(1):
                    print("PERDU:Vous ne pouvez plus gagner")
                    fin_partie=True
            else:
                positionJ1+=valeur
                sequence_en_cours.append("J1_"+str(coups_possibles.pop(coups_possibles.index(valeur)))+"  ")#On enleve de la liste le coup que l'on vient de jouer
                if positionJ2!=0:
                    joueur=2
    else:
        if positionJ2-valeur<=positionJ1 and not croisement:
            positionJ2=positionJ1
            croisement=True
            sequence_en_cours.append("J2_"+str(coups_possibles.pop(coups_possibles.index(valeur)))+"  ")#On enleve de la liste le coup que l'on vient de jouer
            if positionJ1!=distance_totale:
                joueur=1
        else:
            if positionJ2-valeur<0:
                print("ATTENTION:Vous ne pouvez pas jouer ce coup car vous sortez du plateau")
                if not coups_encore_possible(2):
                    print("PERDU:Vous ne pouvez plus gagner")
                    fin_partie=True
            else:
                positionJ2-=valeur
                sequence_en_cours.append("J2_"+str(coups_possibles.pop(coups_possibles.index(valeur)))+"  ")#On enleve de la liste le coup que l'on vient de jouer
                if positionJ1!=distance_totale:
                    joueur=1
    if positionJ1==distance_totale and positionJ2==0:
        print("Vous avez gagné")
        afficher_plateau()
        fin_partie=True
    elif len(coups_possibles)==0:
        print("PERDU : Plus de bornes disponibles")
        afficher_plateau()
        fin_partie=True

def partie(joueur_deb,nouvelle=True):#on donne le joueur qui démarre et si on souhaite modifier le plateur initial et les cartes disponibles après la première partie
    global coups_possibles,positionJ1,positionJ2,joueur,croisement,distance_totale,nb_cartes,plateau,fin_partie
    init_jeu(joueur_deb,nouvelle)
    while not fin_partie:
        afficher_plateau()
        # Affichage des coups possibles
        for i in range(len(coups_possibles)):
            print("Choix",i,":",coups_possibles[i])
        print("C'est au joueur",str(joueur),"de jouer")
        # Attente d'un choix de coup valide
        while True:
            try:
                choix = input("Choix:")
                valeur=coups_possibles[int(choix)]
                break
            except:
                print("Saisie invalide.")
        jouer(valeur)
```
