```python
# On lance une balle depuis un point à 1m du sol
# avec une vitesse de 10 m/s faisant un angle de 35° avec l'horizontale


from matplotlib import pyplot as plt


############################################################
# importation des données à partir du fichier parabole.csv #
############################################################

import csv

fichier = open('parabole2.csv','r') #on ouvre en lecture (r = read) parabole.csv dans "fichier"
FICHIER = csv.reader (fichier,delimiter =';') #on crée un nouveau fichier où
# chaque ligne est une liste constitués d'élements séparés par ";"

#le fichier contient 3 variables t,x,z, on crée 3 listes
Variable0 =[]
Variable1 =[]
Variable2 =[]

for ligne in FICHIER : # on parcourt chaque ligne (= une liste) de FICHIER
    v0 = ligne[0].replace(',','.') # v0 est le premier terme de la liste "ligne". on remplace "," par "."
    Variable0.append(float(v0))
    v1 = ligne[1].replace(',','.')
    Variable1.append(float(v1))
    v2 = ligne[2].replace(',','.')
    Variable2.append(float(v2))

# les élements des listes variable0, variable1 et variable2 sont des chaine de caractères, on les transforme en nombre decimaux grace à float()
# Dans le fichier parabole.csv, la première variable est le temps t, la deuxième est z et la troisième est x

import numpy as np
t = np.array([float(a) for a in Variable0])
z = np.array([float(a) for a in Variable1])
x = np.array([float(a) for a in Variable2])

#             modélisation des équations horaires et cartésiennes
###############################################################################
#                       modélisation de z(t), notée z_t                       #
#                                                                             #
###############################################################################
#                              Travail à faire                                #
# Créer un tableau coef_z_t contenant les coefficients a, b et c du polynôme  #
# de degré 2 modélisant z(t).                                                 #
# avec coef_z_t[0] = a, coef_z_t[1] = b, coef_z_t[2] = c                      #
###############################################################################











###############################################################################
#                              Travail à faire                                #
# Créer un tableau modélisant z(t)                                            #
#                                                                             #
###############################################################################












# modélisation de x(t)
coef_x_t = np.polyfit(t,x,1) #c'est une droite, donc un polynome de degré 1

#creation des coefficients du polynomes (facultatif)
a2 = coef_x_t[0]
b2 = coef_x_t[1]

#on crée le tableau modélisant x(t)
mod_x_t = a2*t+b2

# tant qu'on y est, on modélise l'équation cartésienne !
coef_z_x = np.polyfit(x,z,2)

#creation des coefficients du polynomes (facultatif)
a3 = coef_z_x[0]
b3 = coef_z_x[1]
c3 = coef_z_x[2]

#on crée la modélisation de y(x)
mod_z_x = a3*x**2+b3*x+c3




# Tracés des courbes

plt.figure('z=f(t)') #trace la courbe dans dans une fenetre nommée z=f(t)
plt.plot(t,z,'b+',label ='points expérimentaux') # on affiche les points (+) expérimentaux en bleu
plt.plot(t,mod_z_t,'r-',label ='modélisation degré 2') #on affiche la modélisation en rouge
plt.title('modélisation de z en fonction de t')
plt.xlabel('temps en s')
plt.ylabel('altitude en m')
plt.legend()

plt.figure('x=f(t)') #trace la courbe dans dans une fenetre nommée x=f(t)
plt.plot(t,x,'b+',label ='points expérimentaux') # on affiche les points (+) expérimentaux en bleu
plt.plot(t,mod_x_t,'r-',label ='modélisation degré 2') #on affiche la modélisation en rouge
plt.title('modélisation de x en fonction de t')
plt.xlabel('temps en s')
plt.ylabel('abscisse en m')
plt.legend()

plt.figure('z=f(x)') #trace la courbe dans dans une fenetre nommée z=f(x)
plt.plot(x,z,'b+',label ='points expérimentaux') # on affiche les points (+) expérimentaux en bleu
plt.plot(x,mod_z_x,'r-',label ='trajectoire parabolique') #on affiche la modélisation en rouge
plt.title('modélisation de la trajectoire')
plt.xlabel('abscisse en m')
plt.ylabel('altitude en m')
plt.axis('equal')
plt.legend()


plt.show()

```
