```python
# On lance une balle depuis un point à 1m du sol
# avec une vitesse de 10 m/s faisant un angle de 35° avec l'horizontale


from matplotlib import pyplot as plt


############################################################
# importation des données à partir du fichier parabole.csv #
############################################################

import csv

fichier = open('parabole.csv','r') #on ouvre en lecture (r = read) parabole.csv dans "fichier"
FICHIER = csv.reader (fichier,delimiter =';') #on crée un nouveau fichier où
# chaque ligne est une liste constitués d'élements séparés par ";"

#le fichier contient 3 variables t,z et x, dans cet ordre, on crée 3 listes
t =[]
z =[]
x =[]

##################################################################################################
# Travail à faire : creer la boucle pour remplir les liste t, z et x à partir du fichier importé #
##################################################################################################




############################################
# Création des listes de vitesses vx et vz #
############################################

vx = [None] #on crée un premier élement vide de la liste vx car le 1ier terme n'est pas calculable
vz = [None] # ainsi, les indices des liste de vx, vz, x, z et t coincident

#############################################################################################
# Travail à faire : completer la boucle pour calculer les valeur de la vitesse verticale vz #
#############################################################################################
for i in range(1,len(t)-1) : # len(t)-1 car on ne peut pas calculer v pour le dernier point, d'indice len(t)-1
    vx.append((x[i+1]-x[i-1])/(t[i+1]-t[i-1]))


# Tracé de la trajectoire et des vecteurs vitesse
plt.plot(x,z,"b+-")
#############################################################################################
# Travail à faire : ajouter une boucle pour tracer les vecteurs vitesse, un point sur deux. #
#############################################################################################

plt.title('trajectoire et vecteurs vitesse')
plt.xlabel('abscisses en m')
plt.ylabel('altitude en m')
plt.axis('equal')
plt.grid()
plt.show()

```
