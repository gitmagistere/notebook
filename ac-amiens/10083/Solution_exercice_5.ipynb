```python
# modelisation de l'avancement de la réaction de combustion de l'éthanol
# dans le dioxygène.

# les deux lignes ci-dessous sont à supprimer pour la partie I
nO2i = 25
nC2H6Oi = 8

###########################################
#         partie I : input et print       #
###########################################

# Quel est l'état initial ?

nO2i = float(input('Quelle est la quantité de matière initiale de dioxygène '))
#la réponse à une question est un "mot" (une chaine de caractère type string str
#float transforme ce mot en nombre décimal (int ferait la même chose mais en ne
# gardant que la partie entière

nC2H6Oi = float(input("Quelle est la quantité de matière d'éthanol à l'état initial ? "))
#le texte de la question n'est pas entre '' mais entre "" car le texte contient
# une '
print('Le chimiste a fait réagir ',nO2i,' mol de dioxygène et ',nC2H6Oi," mol d'éthanol")



##########################################
#        partie II : if, else, elif      #
##########################################

x1 = nO2i/3
x2 = nC2H6Oi/1

# Determination du réactif limitant
if x1<x2 :
    xmax = x1
    Rlimitant = 'dioxygène'

elif x1>x2 :        # à completer par le stagiaire pour le cas ou l'ethanol est limitant
    xmax = x2
    Rlimitant = 'éthanol'

else :
    xmax = x1
    Rlimitant ='les deux'


#affichage du résultat
if Rlimitant =='les deux' :
    print('Les deux réactifs sont limitants, nous sommes dans les conditions stoechiométriques, xmax vaut ',round(xmax,1),' mol')
else :
    print('Le réactif limitant est ',Rlimitant,'et xmax vaut',round(xmax,1),' mol')
    # round arrondi la valeur de xmax a 1 chiffre après la virgule

```
