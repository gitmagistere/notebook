# AFFICHER DES GRAPHIQUES AVEC PYTHON et MATPLOTLIB

Il faut importer un module qui permettra d'afficher des graphiques : matplotlib.
On peut le coupler à numpy pour avoir des fonctions mathématiques.
Ces deux modules sont intégrés à Basthon et à Capytale
Vous pouvez aller voir ce site : https://courspython.com/introduction-courbes.html  
On importe en début de programme les modules matplotlib et numpy à l'aide de ces commandes :

```python
import numpy as np
import matplotlib.pyplot as plt
```

On afiiche ensuite les points à l'aide de la commande <code>plt.plot(x,y)</code> où <code>x</code> et <code>y</code> sont les coordonnées du point à placer sur un repère orthonormé. L’instruction plt.plot() permet de tracer des courbes qui relient des points dont les abscisses et ordonnées sont fournies dans des tableaux.

```python
plt.clf() # on efface les courbes précédentes s'il y en a eu
x = [2, 10]
y = [2, 4]
plt.plot(x, y, "r--o")
plt.show() # affiche la figure à l'ecran
```

Le troisième paramètre de plt.plot permet de faire un choix d'affichage : <code>r</code> signifie red, <code>o</code> signifie point rond, <code>--</code> signifie trait pointillé.

Il est utile d'ajouter des légendes aux axes. Il suffit d'ajouter <code>plt.xlabel</code> et <code>plt.ylabel</code>:

```python
plt.clf() # on efface les courbes précédentes s'il y en a eu
x = [2, 10]
y = [2, 4]
plt.plot(x, y, "b-o")
plt.xlabel("abscisses")
plt.ylabel("ordonnees")
plt.show() # affiche la figure à l'ecran
```

On peut réaliser le traçage de la courbe en créant deux listes que l'on alimente dans une boucle, puis on trace...

```python
plt.clf() # on efface les courbes précédentes s'il y en a eu
listex = [] # je crée une liste vide listex
listey = [] # je crée une liste vide listey

for x in range(10): # je fais une boucle où x va prendre les valeurs de 0 à 9
    y = np.sin(x)      # je calcule y comme étant le sinus de x
    listex.append(x) # j'ajoute la valeur x à ma listex
    listey.append(y) # j'ajoute la valeur y à ma listey

x = np.array(listex) # je crée mon tableau pour les x avec la listex
y = np.array(listey) # je crée mon tableau pour les y avec la listey
plt.plot(x, y, "b:o")        # Je place mes points sur le graphique
plt.xlabel("abscisses")
plt.ylabel("ordonnees")

plt.show() # affiche la figure à l'ecran
```
