Il faut importer un module qui permettra d'afficher des graphiques : pylab.

On importe en début de programme le module pylab à l'aide de cette commande :

```python
from pylab import *
```

On afiiche ensuite les points à l'aide de la commande <code>plot(x,y)</code> où <code>x</code> et <code>y</code> sont les coordonnées du point à placer sur un repère orthonormé. L’instruction plot() permet de tracer des courbes qui relient des points dont les abscisses et ordonnées sont fournies dans des tableaux.

```python
x = array([2, 10])
y = array([2, 4])
plot(x, y, "b-o")
show() # affiche la figure à l'ecran
```

Le troisième paramètre de plot permet de faire un choix d'affichage : <code>b</code> signifie blue, <code>-</code> signifie trait continu, et <code>o</code> signifie point rond.

On peut réaliser le traçage de la courbe en créant deux listes que l'on alimente dans une boucle, puis on trace...

```python
listex = [] # je crée une liste vide listex
listey = [] # je crée une liste vide listey

for x in range(10): # je fais une boucle où x va prendre les valeurs de 0 à 9
    y = sin(x)      # je calcule y comme étant le sinus de x
    listex.append(x) # j'ajoute la valeur x à ma listex
    listey.append(y) # j'ajoute la valeur y à ma listey

x = array(listex) # je crée mon tableau pour les x avec la listex
y = array(listey) # je crée mon tableau pour les y avec la listey
plot(x, y)        # Je place mes points sur le graphique

show() # affiche la figure à l'ecran
```
