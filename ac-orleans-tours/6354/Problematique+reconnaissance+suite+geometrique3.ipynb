 Évolution des loyers    

         L’IRL, Indice de Référence des Loyers, sert à fixer les limites des hausses annuelles des loyers que peuvent
         exiger les propriétaires.
        Un propriétaire a suivi, depuis 2008, l’augmentation moyenne de l’IRL chaque année pour 
        la location d’un pavillon. 
        Le tableau suivant regroupe les montants mensuels, en euro, des loyers demandés aux locataires chaque année 
        entre 2008 et 2019 :
    Année	2008	2009	2010	2011	2012	2013	2014	2015	2016	2017	2018	2019
 Montant  	807,15	814,79	821,32	836,38	854,09	863,33	867,66	868,18	868,80	875,78	888,08	900,00
