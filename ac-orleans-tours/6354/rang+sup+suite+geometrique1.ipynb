LOUER A ORLEANS 

             En  2019, un couple a emménagé dans un pavillon à Orléans. Le montant annuel de leur loyer était 
             alors de 10 800 €. 
            L’évolution de l’IRL peut être modélisée par une augmentation moyenne de 1,27 % par an. 
            On considère que cette tendance observée depuis 1998 va se maintenir ainsi durant les années à venir.
            Un pavillon tel que celui que loue le couple se vend à 320 000 €.
