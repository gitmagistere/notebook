## Déclaration d'une fonction

Une fonction : 
* est défini par l'instruction `def`,
* peut prendre un ou plusieurs argument(s),
* peut réaliser une ou plusieurs action(s)...

## Exemples simplistes

### Exemple 1 : triple d'un nombre
La fonction ci-dessous permet de calculer le triple d'un nombre.

```python
def triple(nombre):
  return(nombre*3)
```

L'instruction `return` permet à la fonction de renvoyer un résultat.  

Pour utiliser cette fonction et déterminer le triple de 7, il suffit d'écrire  `triple(7)`

```python
triple(7)
```

### Exemple 2 : Mettre en forme
On définit la fonction `affichage` qui permet d'automatiser la mise en forme d'une partie.

```python
def affichage(numero,titre):
  print("----------------------------------")
  print("Section",numero,":",titre)
  print("----------------------------------")

affichage(1,"Activités")
print("blabla blabla blabla blabla blabla blabla")
print("blabla blabla blabla blabla blabla blabla")
print("blablabla blablabla blabla")
print("blabla blabla blabla blabla blabla blabla")
print("blablabla blablabla blabla")
print()
affichage(2,"Mise en commun")
print("blabla blabla blabla blabla blabla blabla")
print("blabla blabla blabla blabla blabla blabla")
print("blablabla blablabla blabla")
```

Dans la fonction `affichage`, remplacer le mot **section** par le mot **chapitre**, puis cliquer sur Exécuter. 

La modification de la fonction `affichage` s'opère sur l'ensemble du programme.
