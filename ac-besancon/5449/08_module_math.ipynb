## Exemple : calcul d'aire

Créer un programme qui demande le rayon d’un disque et renvoi la valeur de son aire arrondie au dixième.

La fonction `round(VAL,n)` permet d’arrondir la valeur `VAL` à `n` chiffre(s)
après la virgule.

Cependant le nombre π ne fait pas des valeurs de base du langage python. 

```python
print(pi)
```

Il est donc nécessaire d'importer le module `math` pour utiliser la constante `pi`. 

```python
from math import *
print(pi)
```

Le programme demandé est donc 

```python
from math import pi
rayon = int(input("Que vaut le rayon du disque ?"))
print("L’aire du disque est de",round (pi*rayon**2,1))
```
