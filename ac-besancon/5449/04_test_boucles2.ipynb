## Les tests

Les tests permettent le traitement de l'information, ils sont donc essentiels en python (ainsi que dans tout programme informatique).

Pour effectuer un test, il est nécessaire d'avoir des opératoires de comparaisons. 

bases_comparaison.png

### Exemple de test

```python
7 < 0
```

Le test de comparaison ci-dessous renvoi un booléen qui peut prendre deux valeurs : `True` ou `False`

Les tests de comparaison sont utilisés à l'interieur d'une instruction permettant au programme de prendre une décision.

## Les instructions conditionnelles

```python
nombre=7
if (nombre<0):
  print(nombre,"est négatif")
else:
  print(nombre,"est positif")
```

L'instruction `if` indique au programme l'instruction à exécuter en fonction du résultat du test `nombre<0`.

Remarque : Penser à changer la valeur de la variable nombre pour tester le comportement de l'instruction `if`
