### Idée 2

Une seconde idée consiste à faire demander notre prénom à l'aide de l'instruction ***input*** puis d'afficher la réponse précédée du mot bonjour.

L'instruction ***input()*** permet à l'utilisateur de saisir une donnée afin d'interagir avec le progamme

```python
prenom=input("Quel est votre prénom ? ")
print("Bonjour"+prenom)
```
