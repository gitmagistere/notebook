Premier programme

Traditionnellement, les apprentis programmeurs commencent leur carrière en écrivant un programme qui permet d'afficher à l'écran le message Hello World ! Nous n'allons pas déroger à cette tradition.

En langage Python, il suffit d'une seule instruction pour afficher ce message (notez bien que selon le langage utilisé cela peut être plus complexe) :

print("Hello World !")

Recopier cette ligne dans la fenêtre éditeur et lancer l'exécution du programme en cliquant sur run. La console doit afficher « Hello World ! ».

Si c'est le cas, bravo ! Vous venez d'écrire votre premier programme. Il est possible d'utiliser des apostrophes à la place des guillemets : print ('Hello World !')

```
print("Hello World !")
```
