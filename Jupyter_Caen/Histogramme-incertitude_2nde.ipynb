# Tracer un histogramme de donnnées

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Gaussienne

```python
x = np.linspace(0, 20, 100)
```

```python
# moyenne
m = 11.5
# écart-type
s = 1.2
```

```python
y = 1/(s*math.sqrt(2*math.pi)) * np.exp(-(x-m)**2 / (2*s**2))
```

```python
plt.plot(x, y)
plt.text(0.1, 0.25, r'$f=\dfrac{1}{s\sqrt{2\pi}} \times \exp\left({-\dfrac{(x-m)^2}{2s^2}}\right)$')
```
