# Premières fonctions

L’utilisation d’opérateurs mathématiques spécifique nécessite l’utilisation d’une bibliothèque : math. Comme pour la bibliothèque Turtle, la commande 
```python
from math import *
```
permet d'importer l'intégralité de la bibliothèque. 
Une autre manière de faire est d'utiliser la commande 
```python
import math
```
La différence sera alors dans la manière d'appeler les méthodes propres à la bibliothèque.

````python
from math import *
sqrt(4)
````
